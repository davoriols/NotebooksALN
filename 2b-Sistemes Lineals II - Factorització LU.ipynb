## Propietats de l'eliminació Gaussiana

Cada pas de l'eliminació Gaussiana es pot expressar com un producte de matrius:

*   $A^{(k+1)} = G^{(k)} A^{(k)}$
*   $\boldsymbol{b}^{(k+1)} = G^{(k)} \boldsymbol{b}^{(k)}$

per $k=1\ldots,n-1$ on

$$
G^{(k)} =
\begin{pmatrix}
1 &        & \\
  & &\ddots & \\
  & &       & 1 \\
  & &       & -m_{k+1,k} & 1 & & \\
  & &       &  \vdots    &   & &\ddots & \\
  & &       & -m_{n,k}   &   &  &      & 1
\end{pmatrix},
$$
amb $m_{jk} = \cfrac{a_{jk}^{(k)}}{a_{kk}^{(k)}}$ per $j=k+1,\ldots,n$.


&nbsp;

Veiem, en efecte que

$$
\begin{aligned}
 G^{(1)}A^{(1)} &=
\begin{pmatrix}
1 \\
-m_{21}& 1\\
\vdots & & &\ddots\\
-m_{n1} & & & & 1
\end{pmatrix}
\cdot
\begin{pmatrix}
a_{11}^{(1)} & a_{12}^{(1)} & \cdots & a_{1n}^{(1)}\\
a_{21}^{(1)} & a_{22}^{(1)} & \cdots & a_{2n}^{(1)}\\
\vdots & \vdots &  & \vdots\\
a_{n1}^{(1)} & a_{n2}^{(1)} & \cdots & a_{nn}^{(1)}\\
\end{pmatrix}\\[1.5ex]
& =
\begin{pmatrix}
a_{11}^{(1)} & a_{12}^{(1)} & \cdots & a_{1n}^{(1)}\\
0 & a_{22}^{(1)} - m_{21}a_{12}^{(1)} & \cdots &a_{2n}^{(1)} - m_{21}a_{1n}^{(1)}\\
\vdots & \vdots &  & \vdots\\
0 & a_{n2}^{(1)} - m_{n1}a_{12}^{(1)} & \cdots & a_{nn}^{(1)} - m_{n1}a_{1n}^{(1)}
\end{pmatrix}\\[1.5ex]
& =
\begin{pmatrix}
a_{11}^{(1)} & a_{12}^{(1)} & \cdots & a_{1n}^{(1)}\\
0 & a_{22}^{(2)} & \cdots & a_{2n}^{(2)}\\
\vdots & \vdots &  & \vdots\\
0 & a_{n2}^{(2)} & \cdots & a_{nn}^{(2)}\\
\end{pmatrix}\\[1.5ex]
& = A^{(2)}
\end{aligned}
$$


&nbsp;

i, en general:

$$
\begin{aligned}
 G^{(k)}A^{(k)}
& =
\begin{pmatrix}
1 &        & \\
  & &\ddots & \\
  & &       & 1 \\
  & &       & -m_{k+1,k} & 1 & & \\
  & &       &  \vdots    &   & &\ddots & \\
  & &       & -m_{n,k}   &   &  &      & 1
\end{pmatrix}
\cdot
\begin{pmatrix}
a_{1,1}^{(1)} & a_{1,2}^{(1)} & \cdots & a_{1,k}^{(1)} & a_{1,k+1}^{(1)} & \cdots & a_{1,n}^{(1)} \\
             & a_{2,2}^{(2)} & \cdots & a_{2,k}^{(2)} & a_{2,k+1}^{(2)} & \cdots & a_{2,n}^{(2)} \\
             &              & \ddots & \vdots   & \vdots    &        & \vdots       \\
             &              &        & a_{k,k}^{(k)} & a_{k,k+1}^{(k)} & \cdots & a_{k,n}^{(k)} \\
             &              &        & a_{k+1,k}^{(k)} & a_{k+1,k+1}^{(k)} & \cdots & a_{k+1,n}^{(k)} \\
             &              &        & \vdots       &    \vdots &    & \vdots       \\
             &              &        & a_{n,k}^{(k)} & a_{n,k+1}^{(k)} & \cdots & a_{n,n}^{(k)} \\
\end{pmatrix}\\[2.25ex]
& =
\begin{pmatrix}
a_{1,1}^{(1)} & a_{1,2}^{(1)} & \cdots & a_{1,k}^{(1)} & a_{1,k+1}^{(1)} & \cdots & a_{1,n}^{(1)} \\
             & a_{2,2}^{(2)} & \cdots & a_{2,k}^{(2)} & a_{2,k+1}^{(2)} & \cdots & a_{2,n}^{(2)} \\
             &              & \ddots & \vdots   & \vdots    &        & \vdots       \\
             &              &        & a_{k,k}^{(k)} & a_{k,k+1}^{(k)} & \cdots & a_{k,n}^{(k)} \\
             &              &        &   & a_{k+1,k+1}^{(k)} -m_{k+1,k}a_{k,k+1}^{(k)} & \cdots & a_{k+1,n}^{(k)} -m_{k+1,k}a_{k,n}^{(k)}  \\
             &              &        &   &    \vdots &    & \vdots       \\
             &              &        &   & a_{n,k+1}^{(k)} -m_{n,k}a_{k,k+1}^{(k)}& \cdots & a_{n,n}^{(k)} -m_{n,k}a_{k,n}^{(k)}\\
\end{pmatrix}\\[2.25ex]
& =
\begin{pmatrix}
a_{1,1}^{(1)} & a_{1,2}^{(1)} & \cdots & a_{1,k}^{(1)} & a_{1,k+1}^{(1)} & \cdots & a_{1,n}^{(1)} \\
             & a_{2,2}^{(2)} & \cdots & a_{2,k}^{(2)} & a_{2,k+1}^{(2)} & \cdots & a_{2,n}^{(2)} \\
             &              & \ddots & \vdots   & \vdots    &        & \vdots       \\
             &              &        & a_{k,k}^{(k)} & a_{k,k+1}^{(k)} & \cdots & a_{k,n}^{(k)} \\
             &              &        &   & a_{k+1,k+1}^{(k+1)} & \cdots & a_{k+1,n}^{(k+1)} \\
             & & & &    \vdots &    & \vdots       \\
             & & & & a_{n,k+1}^{(k+1)} & \cdots & a_{n,n}^{(k+1)}
\end{pmatrix}\\[2.25ex]
& =
A^{(k+1)}
\end{aligned}
$$




Per tant, tenim:

$$
A^{(n)}
=
G^{(n-1)} A^{(n-1)} = \underbrace{G^{(n-1)}\cdot \ldots \cdot G^{(1)}}_{\hat{L}} \cdot A^{(1)} = \hat{L} \cdot A^{(1)}
$$

on $\hat{L}$ és una matriu triangular inferior.

Observem dues coses:

1.   $A^{(n)} = U$ és una matriu triangular superior
2.   $ L = \hat{L}^{-1}$ és una matriu traingular inferior

Així podem expressar $A$ com el producte d'una matriu triangular inferior i una matriu triangular superior:

$$A = L\cdot U .$$


&nbsp;


### Expressió de $L$:

Notem que

$$
\left(G^{(k)}\right)^{-1} =
\begin{pmatrix}
1 &        & \\
  & &\ddots & \\
  & &       & 1 \\
  & &       & m_{k+1,k} & 1 & & \\
  & &       &  \vdots    &   & &\ddots & \\
  & &       & m_{n,k}   &   &  &      & 1
\end{pmatrix},
$$

 i que, per definició, sabem que $L = \hat{L}^{-1}$, així:

$$
\begin{aligned}
L &= \hat{L}^{-1} = \left( G^{(n-1)}\cdot \ldots \cdot G^{(1)}\right)^{-1} = \left(G^{(1)}\right)^{-1} \cdot \ldots \cdot \left(G^{(n-1)}\right)^{-1}\\
& = \begin{pmatrix}
1 &        & \\
m_{2,1}  & 1 & \\
\vdots & \ddots & \ddots \\
m_{n,1} & \cdots &m_{n,n-1} & 1
\end{pmatrix}
\end{aligned}
$$

és a dir, $L$ és una matriu diagonal amb uns a la diagonal i que els seus elements són els multiplicadors de l'eliminació Gaussiana.

&nbsp;

## Factorització $LU$:

Així, únicament hem de guardar els multiplicadors dins la matriu $L$.

Observem que com la matriu $L$ conté uns a la diagonal, podem guardar les matrius $L$ i $U$ en una mateixa matriu $M$, tal que $M = L+ U -Id$, és a dir:

$M =
\begin{pmatrix}
a_{1,1}^{(1)} & a_{1,2}^{(1)} & \cdots & a_{1,n-1}^{(1)} & a_{1,n}^{(1)} \\
m_{2,1}             & a_{2,2}^{(2)} & \cdots & a_{2,n-1}^{(2)} & a_{2,n}^{(2)} \\
 \vdots            &  \ddots            & \ddots & \vdots   & \vdots   \\
 m_{n-1,1}            &   \cdots           & m_{n-1,n-2}        & a_{n-1,n-1}^{(n-1)} & a_{n-1,n-1}^{(n-1)}\\
 m_{n,1}            &   \cdots           & m_{n,n-2}        & m_{n,n-1} & a_{n,n}^{(n)}\\
\end{pmatrix}
$

&nbsp;

#### Nombre d'operacions:

El nombre d'operacions de l'algorisme és
$$
\frac{2}{3}n^3 - \frac{n^2}{2} - \frac{n}{6} \approx \frac{2}{3}n^3
$$

### Resolució de sistemes lineals utilitzant la factorització $LU$

Per resoldre un sistema lineal $A\boldsymbol{x} = \boldsymbol{b}$ utilitzant la factorització $LU$ farem el següent procediment:

1. Factoritzem la matriu $A=LU$

2. Resolem el sistema $L\boldsymbol{y} = \boldsymbol{b}$ utilitzant l'algorisme de substitució endavant.

3. Resolem el sistema $U \boldsymbol{x} = \boldsymbol{y}$ utilitzant l'algorisme de substitució endarrere.

## Mètode de Doolittle

El mètode de Dollittle ens proporciona un mètode per al càlcul de la factorització $LU$ que no utilitza l'eliminació Gaussiana. El procediment per fer-lo és:

$$
\left\{
\begin{aligned}
& u_{kj} = a_{kj} - \sum_{r=1}^{k-1} l_{kr}u_{rj} \quad \text{ per }j=k,\ldots,n\\
& l_{kk} = 1\\
& l_{ik} = \frac{1}{u_{kk}}\left(a_{ik} - \sum_{r=1}^{k-1}l_{ir}u_{rk}\right) \quad \text{ per } i=k+1,\ldots,n
\end{aligned}
\right.
$$

per $k=1,\ldots,n$.

Observem d'igual forma que per l'algorisme funcioni necessitem que $u_{kk}$ sigui diferent de $0$.


### Algorisme de factorització de Doolittle



L'algorisme de factorització de Doolittle és el següent:

$
\begin{aligned}
&\texttt{for } k = 1 \div n \\
&\phantom{for } \texttt{for } j = k \div n \\
&\phantom{for for } u_{kj} \leftarrow a_{kj} \\
&\phantom{for for } \texttt{for } r = 1 \div k-1\\
&\phantom{for for for}  u_{kj} \leftarrow u_{kj} - l_{kr}u_{rj}\\
&\phantom{for for} \texttt{end for}\\
&\phantom{for } \texttt{end for}\\
&\phantom{for } l_{kk} \leftarrow 1\\
&\phantom{for } \texttt{for } i = k +1 \div n \\
&\phantom{for for } l_{ik} \leftarrow a_{ik} \\
&\phantom{for for } \texttt{for } r = 1 \div k-1\\
&\phantom{for for for}  l_{ik} \leftarrow l_{ik} - l_{ir}u_{rk}\\
&\phantom{for for} \texttt{end for}\\
&\phantom{for for } l_{ik} \leftarrow l_{ik}/u_{kk}\\
&\phantom{for } \texttt{end for}\\
&\texttt{end for}
\end{aligned}
$
