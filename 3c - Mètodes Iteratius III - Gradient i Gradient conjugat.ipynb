# Mètode del Gradient i Mètode del Gradient Conjugat

## Conceptes inicials:


A Càlcul diferencial heu vist (o esteu a punt de veure) com podem caracteritzar els extrems relatius d'una funció $f:U \subset \mathbb{R}^n\rightarrow \mathbb{R}$ amb $U$ obert. Recordem-ho ràpidament:

##### **Definició:** *Gradient*

Donada una funció real de varies variables $f(x_1,x_2,\ldots,x_n)$ de la que existeixen les derivades parcials definim el seu gradient com

$$
\nabla f(\boldsymbol{x}) := \begin{pmatrix}\frac{\partial f(\boldsymbol{x})}{\partial x_1}\\ \frac{\partial f(\boldsymbol{x})}{\partial x_2}\\ \vdots \\\frac{\partial f(\boldsymbol{x})}{\partial x_n}\end{pmatrix}.
$$

##### **Definició:** *Punt crític*

Diem que $f: U \subset \mathbb{R}^n \rightarrow \mathbb{R}$ té un punt crític a $\boldsymbol{x}_0\in U$ si $f$ és diferenciable en $\boldsymbol{x}_0$ i $\nabla f(\boldsymbol{x}_0) = \boldsymbol{0}$.


##### **Definició:** *Matriu Hessiana*

Donada un funció real de varies variables $f(x_1,x_2,\ldots,x_n)$ de la que existeixen les segones derivades parcials definim la seva matriu Hessiana com:

$$
    Hf(\boldsymbol{x}) = \left(\frac{\partial^2 f(\boldsymbol{x})}{\partial x_i \partial x_j}\right)_{i,j} = \begin{pmatrix} \frac{\partial^2 f(\boldsymbol{x})}{\partial x_1 \partial x_1} &  \frac{\partial^2 f(\boldsymbol{x})}{\partial x_1 \partial x_2} & \cdots & \frac{\partial^2 f(\boldsymbol{x})}{\partial x_1 \partial x_n}\\ \frac{\partial^2 f(\boldsymbol{x})}{\partial x_2 \partial x_1} &  \frac{\partial^2 f(\boldsymbol{x})}{\partial x_2 \partial x_2} & \cdots & \frac{\partial^2 f(\boldsymbol{x})}{\partial x_2 \partial x_n}\\ \vdots & \vdots & \ddots & \vdots\\ \frac{\partial^2 f(\boldsymbol{x})}{\partial x_n \partial x_1} &  \frac{\partial^2 f(\boldsymbol{x})}{\partial x_n \partial x_2} & \cdots & \frac{\partial^2 f(\boldsymbol{x})}{\partial x_n \partial x_n} \end{pmatrix}.
$$

#### **Teorema:** *Caracterització d'extrems realtius*

Sigui $\boldsymbol{x}_0\in\mathbb{R}^n$ i $f\in C^2(B(\boldsymbol{x}_0,\varepsilon))$ amb $\varepsilon > 0$. Suposem que $\boldsymbol{x}_0$ és un punt crític de $f$ (és a dir $\nabla f(\boldsymbol{x}_0) = \boldsymbol{0}$). Aleshores:

1.   Si $Hf(\boldsymbol{x}_0)$ és definida positiva, $f$ té un mínim local a $\boldsymbol{x}_0$.
2.   Si $Hf(\boldsymbol{x}_0)$ és definida negativa, $f$ té un màxim local a $\boldsymbol{x}_0$.

&nbsp;

### Aplicació al cas particular de sistemes lineals

Considerem el sistema lineal $A\boldsymbol{x} = \boldsymbol{b}$ amb $A$ matriu quadrada **simètrica i definida positiva**.

Observem que determinar el vector $\boldsymbol{x}$ és equivalent a determinar el valor de $\boldsymbol{x}$ tal que $A\boldsymbol{x} - \boldsymbol{b} = \boldsymbol{0}$.

Tenint això en ment, definim la funció:

$$
    f(\boldsymbol{x}) := \frac{1}{2}\boldsymbol{x}^\top A\boldsymbol{x} - \boldsymbol{x}^\top \boldsymbol{b}.
$$

Observem que:

* $\nabla f(\boldsymbol{x}) = A\boldsymbol{x} - \boldsymbol{b}.$

* $Hf(\boldsymbol{x})= A.$

&nbsp;

Així doncs, fixem-nos que si $\nabla f(\boldsymbol{x}) = A\boldsymbol{x} - \boldsymbol{b} = \boldsymbol{0}$, com $A$ és definida positiva tenim que el vector $\boldsymbol{x}$ que és solució del sistema lineal és un mínim de la funció $f$.

#### **Comentari 1:**

Els mètodes del gradient i del gradient conjugat que veurem són mètodes que també serveixen per problemes de minimització de funcions més complexes. Ara bé, pel cas particulars de funcions
$$
    f(\boldsymbol{x}) = \frac{1}{2}\boldsymbol{x}^\top A\boldsymbol{x} - \boldsymbol{x}^\top \boldsymbol{b}.
$$
amb $A$ una matriu simètrica i definida positiva podem trobar formulacions més òptimes computacionalment d'ambdos mètodes.

#### **Comentari 2:**

Pel cas de sistemes lineals $\tilde{A}\boldsymbol{x} = \tilde{\boldsymbol{b}}$ amb $\tilde{A}\in \mathcal{M}_{m\times n}(\mathbb{R})$ amb $m\geq n$ i de rang $n$ podem definir el sistema equivalent:
$$
    \underbrace{\tilde{A}^\top \tilde{A}}_{A} \boldsymbol{x} = \underbrace{\tilde{A}^\top\tilde{\boldsymbol{b}}}_{\boldsymbol{b}}
$$

on ara la matriu del sistema $A = \tilde{A}^\top \tilde{A}$ és simètrica i definida positiva.

Això és equivalent a voler minimitzar la funció:

$$
\begin{aligned}
    g(\boldsymbol{x}) &= \frac{1}{2} \|\tilde{A}\boldsymbol{x} - \tilde{\boldsymbol{b}}||_2^2\\ &= \frac{1}{2}\left(\tilde{A}\boldsymbol{x} - \tilde{\boldsymbol{b}}\right)^\top \left(\tilde{A}\boldsymbol{x} - \tilde{\boldsymbol{b}}\right)\\ & = \frac{1}{2}\left(\boldsymbol{x}^\top\tilde{A}^\top - \tilde{\boldsymbol{b}}^\top\right)\left(\tilde{A}\boldsymbol{x} - \tilde{\boldsymbol{b}}\right)\\ & = \frac{1}{2}\left(\boldsymbol{x}^\top\tilde{A}^\top\tilde{A}\boldsymbol{x} - \boldsymbol{x}^\top\tilde{A}^\top\tilde{\boldsymbol{b}} - \tilde{\boldsymbol{b}}^\top\tilde{A}\boldsymbol{x} + \tilde{\boldsymbol{b}}^\top\tilde{\boldsymbol{b}}\right)\\
    & = \frac{1}{2}\left(\boldsymbol{x}^\top\tilde{A}^\top\tilde{A}\boldsymbol{x} - 2\boldsymbol{x}^\top\tilde{A}^\top\tilde{\boldsymbol{b}} + \tilde{\boldsymbol{b}}^\top\tilde{\boldsymbol{b}}\right)\\
    & = \frac{1}{2}\boldsymbol{x}^\top\underbrace{\tilde{A}^\top\tilde{A}}_{A}\boldsymbol{x} - \boldsymbol{x}^\top\underbrace{\tilde{A}^\top\tilde{\boldsymbol{b}}}_{\boldsymbol{b}} + \underbrace{\frac{1}{2}\tilde{\boldsymbol{b}}^\top\tilde{\boldsymbol{b}}}_{\text{constant}}\\
\end{aligned}    
$$

i, per tant:

* $\nabla g(\boldsymbol{x}) = \tilde{A}^\top\tilde{A}\boldsymbol{x} - \tilde{A}^\top\tilde{\boldsymbol{b}} = A\boldsymbol{x}-\boldsymbol{b}$
* $Hg(\boldsymbol{x})= \tilde{A}^\top\tilde{A} = A$

## Mètode del Gradient:

#### **Idea intuïtiva del mètode general**

Donada la funció $F(\boldsymbol{x})$ idea del mètode consisteix en definir la seqüència $\{\boldsymbol{x}_n\}_{n\geq 0}$ tals que

$$
    F(\boldsymbol{x}_0) \geq F(\boldsymbol{x}_1) \geq \ldots,
$$

de manera que el límit de la successió correspongui amb un mínim (local) de $F$, tot utilitzant el procediment:

$$
    \boldsymbol{x}_{k+1} = \boldsymbol{x}_k - \alpha_k \nabla F(\boldsymbol{x}_k),
$$

on els $\alpha_k> 0$ són coeficients a determinar.

&nbsp;

Fixem-nos que estem modificant el punt $\boldsymbol{x}_k$ en la direcció de màxim descens de $F$ en $\boldsymbol{x}_k$, és a dir, $-\nabla F(\boldsymbol{x}_k)$.

#### **Mètode per sistemes lineals**

Recordem que pel cas de sistemes lineals $A\boldsymbol{x} = \boldsymbol{b}$ amb $A$ simètrica i definida positiva, tenim:
$$
f(\boldsymbol{x}) := \frac{1}{2}\boldsymbol{x}^\top A\boldsymbol{x} - \boldsymbol{x}^\top \boldsymbol{b},
$$
i, per tant:
$$\nabla f(\boldsymbol{x}) = A\boldsymbol{x} - \boldsymbol{b}.$$

Així, podem escriure el mètode com:

$$
\left\{
\begin{aligned}
& \boldsymbol{x}^{(0)} \in \mathbb{R}^n,\\
& \boldsymbol{x}^{(k+1)} = \boldsymbol{x}^{(k)} + \alpha_k(\boldsymbol{b}- A\boldsymbol{x}^{(k)}),\quad \text{ per } k\geq 0.
\end{aligned}    
\right.    
$$


&nbsp;

En aquest cas, el valor $\alpha_k$ òptim, ve donat per:

$$
\alpha_k = \frac{\left(A\boldsymbol{x}^{(k)} - \boldsymbol{b}\right)^\top \left(A\boldsymbol{x}^{(k)} - \boldsymbol{b}\right)}{\left(A\boldsymbol{x}^{(k)} - \boldsymbol{b}\right)^\top A\left(A\boldsymbol{x}^{(k)} - \boldsymbol{b}\right)}.
$$

D'aquesta forma, l'algorisme, donada la matriu $A$ i el terme independent $\boldsymbol{b}$ del sistema, l'aproximació inicial $\boldsymbol{x}^{(0)}$ de la solució i un valor de la tolerància $\varepsilon$ amb el que volem la solució farem:

$
\begin{array}{lll}
&\boldsymbol{r}_0 \leftarrow \boldsymbol{b}-A\boldsymbol{x}^{(0)}& \,(\text { observem que } \boldsymbol{r}_0 = - \nabla f(\boldsymbol{x}^{(0)})\,\,)\\
&k \leftarrow 0\\
&\texttt{while } \|\boldsymbol{r}_k\| \geq \varepsilon\\
&\phantom{for }\alpha_k \leftarrow \cfrac{\boldsymbol{r}_k^\top\boldsymbol{r}_k}{\boldsymbol{r}_k^\top A \boldsymbol{r}_k}\\
&\phantom{for } \boldsymbol{x}^{(k+1)} \leftarrow \boldsymbol{x}^{(k)} + \alpha_k \boldsymbol{r}_k& \,(\text { observem que } \boldsymbol{r}_{k+1} = - \nabla f(\boldsymbol{x}^{(k+1)})\,\,)\\
&\phantom{for } \boldsymbol{r}_{k+1} \leftarrow \boldsymbol{b}-A\boldsymbol{x}^{(k+1)} \\
&\phantom{for }k \leftarrow k+1\\
&\texttt{end}
\end{array}
$

&nbsp;

El mètode anterior es pot adaptar per tal d'evitar calcular el producte $A\boldsymbol{x}^{(k+1)}$ tot reutilitzant el càlcul de $A\boldsymbol{r}_k$. Ja que, tenim:

$$
\begin{aligned}
    \boldsymbol{r}_{k+1} - \boldsymbol{r}_k &= (\boldsymbol{b}- A\boldsymbol{x}^{(k+1)}) - (\boldsymbol{b}- A\boldsymbol{x}^{(k)})\\
    & = - A\boldsymbol{x}^{(k+1)} + A\boldsymbol{x}^{(k)}\\
    & = - A(\boldsymbol{x}^{(k+1)}- \boldsymbol{x}^{(k)})\\
    & = - A(\alpha_k \boldsymbol{r}_k)\\
    & = - \alpha_k A \boldsymbol{r}_k
\end{aligned}    
$$

i, per tant:

$$
    \boldsymbol{r}_{k+1} = \boldsymbol{r}_{k} - \alpha_k A \boldsymbol{r}_k
$$


&nbsp;

D'aquesta manera, l'algorisme esdevé:

$
\begin{array}{ll}
&\boldsymbol{r}_0 \leftarrow \boldsymbol{b}-A\boldsymbol{x}^{(0)}\\
&k \leftarrow 0 \\
&\texttt{while } \|\boldsymbol{r}_k\| \geq \varepsilon\\
&\phantom{for }\alpha_k \leftarrow \cfrac{\boldsymbol{r}_k^\top\boldsymbol{r}_k}{\boldsymbol{r}_k^\top A \boldsymbol{r}_k}\\
&\phantom{for } \boldsymbol{x}^{(k+1)} \leftarrow \boldsymbol{x}^{(k)} + \alpha_k \boldsymbol{r}_k\\
&\phantom{for } \boldsymbol{r}_{k+1} \leftarrow \boldsymbol{r}_k-\alpha_k A\boldsymbol{r}_k \\
&\phantom{for }k \leftarrow k+1\\
&\texttt{end}
\end{array}
$

## Mètode del Gradient conjugat

El mètode del gradient conjugat és una modificació del mètode anterior que permet una convergència molt més ràpida.

Podeu trobar la construcció del mètode vista a teoria a *Numerical Linear Algebra, Grégoire Allaire, Sidi Mahmoud Kaber, Springer, 2008*.

##### **Remarca important del mètode:**

* El mètode del gradient conjugat convergeix cap a la solució en $\leq n$ iteracions. De fet, podem pensar el mètode com un mètode directe d'ortogonalització, però en general pararem el procediment abans de fer totes les iteracions necessàries.

#### **Algorisme del gradient conjugat**

$
\begin{aligned}
&\boldsymbol{r}_0 \leftarrow \boldsymbol{b} - A\boldsymbol{x}^{(0)}\\
& \boldsymbol{p}_0 \leftarrow\boldsymbol{r}_0\\
& k \leftarrow 0\\
&\texttt{while } \|\boldsymbol{r}^{(k)}\| \geq \varepsilon\\
&\phantom{for }\alpha_k \leftarrow \cfrac{\boldsymbol{r}_k^\top\boldsymbol{r}_k}{\boldsymbol{p}_k^\top A \boldsymbol{p}_k}\\
&\phantom{for } \boldsymbol{x}^{(k+1)} \leftarrow \boldsymbol{x}^{(k)} + \alpha_k \boldsymbol{p}_k\\
&\phantom{for } \boldsymbol{r}_{k+1} \leftarrow \boldsymbol{r}_k-\alpha_k A\boldsymbol{p}_k \\
&\phantom{for } \beta_k \leftarrow \cfrac{\boldsymbol{r}_{k+1}^\top\boldsymbol{r}_{k+1}}{\boldsymbol{r}_k^\top\boldsymbol{r}_k} \\
&\phantom{for } \boldsymbol{p}_{k+1}\leftarrow \boldsymbol{r}_{k+1} + \beta_k\boldsymbol{p}_k\\
&\phantom{for }k \leftarrow k+1\\
&\texttt{end}
\end{aligned}
$

Si introduïm les variables:

$$
\left\{
\begin{aligned}
&\boldsymbol{y}_k = A\boldsymbol{p}_k\\
&\gamma_k = \boldsymbol{r}_k^\top\boldsymbol{r}_k
\end{aligned}
\right.
$$

veiem clarament que podem optimitzar l'espai en memòria necessari per anar iterant el mètode.

&nbsp;

$
\begin{aligned}
&\boldsymbol{r}_0 \leftarrow \boldsymbol{b} - A\boldsymbol{x}^{(0)}\\
& \boldsymbol{p}_0 \leftarrow\boldsymbol{r}_0\\
& k \leftarrow 0\\
& \gamma_0 = \boldsymbol{r}_0^\top\boldsymbol{r}_0\\
&\texttt{while } \gamma_k \geq \varepsilon^2\\
&\phantom{for }\boldsymbol{y}_k \leftarrow A\boldsymbol{p}_k \\
&\phantom{for }\alpha_k \leftarrow \cfrac{\gamma_k}{\boldsymbol{p}_k^\top \boldsymbol{y}_k}\\
&\phantom{for } \boldsymbol{x}^{(k+1)} \leftarrow \boldsymbol{x}^{(k)} + \alpha_k \boldsymbol{p}_k\\
&\phantom{for } \boldsymbol{r}_{k+1} \leftarrow \boldsymbol{r}_k-\alpha_k \boldsymbol{y}_k \\
&\phantom{for } \beta_k \leftarrow \cfrac{\boldsymbol{r}_{k+1}^\top\boldsymbol{r}_{k+1}}{\gamma_k} \\
&\phantom{for }\gamma_{k+1} \leftarrow \boldsymbol{r}_{k+1}^\top\boldsymbol{r}_{k+1}\\
&\phantom{for } \boldsymbol{p}_{k+1}\leftarrow \boldsymbol{r}_{k+1} + \beta_k\boldsymbol{p}_k\\
&\phantom{for }k \leftarrow k+1\\
&\texttt{end}
\end{aligned}
$
