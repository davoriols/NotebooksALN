# Factorització $QR$

Donada una matriu $A\in\mathcal{M}_{m\times n}$ amb $m\geq n$ i rang màxim (és a dir, $\text{rang}(A) = n$), podem calcular la seva factorització $QR$. On:

* $Q$ és una matriu $m\times n$ ortonormal per columnes.

* $R$ és una matriu quadrada $n\times n$ triangular superior.

&nbsp;




**Observació:** Recordem que dos vectors $\boldsymbol{v}_1$ i $\boldsymbol{v}_2$ són ortonormals si són *ortogonals* ($\boldsymbol{v}_1\cdot\boldsymbol{v}_2 = 0$) i *unitaris* ($||\boldsymbol{v}_1|| = ||\boldsymbol{v}_2|| = 1$). Així, si denotem per $\boldsymbol{q}_i$ per $i=1,\ldots,n$ els vectors columna de $Q$, tenim que:
$$
  \boldsymbol{q}_i \cdot \boldsymbol{q}_j = \delta_{ij} = \left\{\begin{aligned}& 1\qquad\text{si }i=j,\\ &0\qquad\text{si } i\neq j.\end{aligned}\right.
$$

($\delta_{ij}$ rep el nom de delta de Kronecker).


&nbsp;

**Observació 2:** La pròpia definició de matriu ortonormal ens implica que $Q^\top Q = Id$.

&nbsp;

**Observació 3:** Si $m=n$ aleshores:

* $Q^\top = Q^{-1}$.

* $\det (Q) = \pm 1$ i, per tant, $\det (A) = \pm \det(R)$.

#### Cas $m$ $=$ $n$:

Suposem que volem resoldre el sistema lineal $A\boldsymbol{x} = \boldsymbol{b}$ mitjançant la factorització $QR$ de $A$. Aleshores, tindriem:

$$
\begin{array}{llll}
A\boldsymbol{x} = \boldsymbol{b} & \Longleftrightarrow QR\boldsymbol{x} = \boldsymbol{b}& & \text{(}A=QR\text{)}\\
& \Longleftrightarrow Q^\top QR\boldsymbol{x} = Q^\top\boldsymbol{b} &\quad&\text{(multipliquem per }Q^\top\text{)}\\
& \Longleftrightarrow R\boldsymbol{x} = Q^\top\boldsymbol{b}& & \text{(}Q^\top Q = Id\text{)}\\
\end{array}
$$

que és un sistema triangular superior que podem resoldre utilitzant la substitució endarrere.


&nbsp;


#### Cas $m$ $>$ $n$:

Podriem fer exactament el mateix procediment que hem fet pel cas anterior:
$$
\begin{array}{llll}
A\boldsymbol{x} = \boldsymbol{b} & \Longleftrightarrow QR\boldsymbol{x} = \boldsymbol{b}& & \text{(}A=QR\text{)}\\
& \Longleftrightarrow Q^\top QR\boldsymbol{x} = Q^\top\boldsymbol{b} &\quad&\text{(multipliquem per }Q^\top\text{)}\\
& \Longleftrightarrow R\boldsymbol{x} = Q^\top\boldsymbol{b}& & \text{(}Q^\top Q = Id\text{)}\\
\end{array}
$$

que és un sistema triangular superior que podem resoldre utilitzant la substitució endarrere.

&nbsp;

Però...

**El sistema** $A\boldsymbol{x} = \boldsymbol{b}$ **té sempre solució?**

Resposta...

**NO, en general el sistema seria incompatible.**

## El problema dels mínims quadrats

En general, el sistema anterior no té solució, a no ser que $\boldsymbol{b}$ sigui combinació lineal de les columnes de $A$. Així doncs, l'objectiu no serà trobar la solució del problema, sinó trobar el vector $\boldsymbol{x}\in\mathbb{R}^n$ més proper possible a la solució.
És a dir, l'objectiu serà minimitzar la norma del residu $\boldsymbol{r} = A\boldsymbol{x} - \boldsymbol{b}$.

La norma que s'utilitza per resoldre aquest problema condiciona el resultat però, en el cas de la norma euclídea, trobar el vector $\boldsymbol{x}$ que minimtza $||A\boldsymbol{x} - \boldsymbol{b}||$ es coneix com a *aproximació per mínims quadrats*.



La propietat més important del vector $\boldsymbol{r}$ òptim és que pertany al nucli de la matriu $A^\top$, és a dir:

$$
A^\top\boldsymbol{r} = \boldsymbol{0}.
$$

Utilitzant la definició de $\boldsymbol{r}$ arribem a l'expressió

$$
A^\top (A\boldsymbol{x} - \boldsymbol{b}) = \boldsymbol{0},
$$

que és equivalent a

$$
A^\top A\boldsymbol{x} = A^\top\boldsymbol{b},
$$

sistema que es coneix com a **equacions normals**.
Notem que, en essència, el que estem fent és projectar el vector $\boldsymbol{b}$ sobre l'espai vectorial generat pels vectors columna de la matriu $A$.



Si la matriu inicial $A$ té rang màxim (tal i com estem suposant), aleshores la matriu $A^\top A$ és simètrica i definida positiva, i per tant podríem pensar que seria raonable aplicar una descomposició de Cholesky $A^\top A = \mathcal{L}\mathcal{L}^\top$. Ara bé, és senzill trobar exemples de matrius $A$ de rang màxim que en aritmètica de coma flotant proporcionen $A^\top A$ propera a singular, cosa que fa que es produeixi un error numèric més gran.

Per evitar aquest problema podem utilitzar la factorització $QR$:



$$
\begin{array}{llll}
A^\top A\boldsymbol{x} = A^\top\boldsymbol{b}
  &\Longleftrightarrow (QR)^\top (QR)\boldsymbol{x} = (QR)^\top\boldsymbol{b} &\quad& \text{(}A=QR\text{)}\\
   &\Longleftrightarrow R^\top Q^\top QR\boldsymbol{x} = R^\top Q^\top\boldsymbol{b} & & \big((QR)^\top = R^\top Q^\top \big)\\
   &\Longleftrightarrow Q^\top QR\boldsymbol{x} = Q^\top\boldsymbol{b}&\quad& \text{(}R\text{ quadrada i rang màxim)}\\
   &\Longleftrightarrow R\boldsymbol{x} = Q^\top\boldsymbol{b}&\quad& \text{(}Q^\top Q = Id\text{)}\\
\end{array}
$$


##### *Exemple:*

*Resolució de les equacions normals utilitzant els diferents mètodes:*

```python
import numpy as np
from scipy.linalg import cholesky, solve_triangular, qr, qr_multiply


# Valors de m i n:
m = 500
n = 300

# Matriu del sistema, mida m x n, el generem aleatòriament
A = np.random.rand(m, n)

# Vector de termes independents, mida m, el generem aleatòriament
b = np.random.rand(m)



######################### RESOLUCIÓ PER CHOLESKY ###############################

# 1) Calculem les matrius A^T A i el vector A^T b
AtA = A.T@A
Atb = A.T@b

# 2) Descomposició Cholesky de la matriu A^T A
L = cholesky(AtA, lower=True)
#L = np.linalg.cholesky(AtA) # També es pot fer amb cholesky de numpy

# 3) Resolem el sistem triangular inferior Ly = Atb per obtenir el valor de y
y = solve_triangular(L, Atb, lower=True)

# 4) Resolem el sistema triangular superior L^T x = y, per obtenir la solució x
x = solve_triangular(L.T, y)



############################### RESOLUCIÓ PER QR ###############################

# 1) Calculem la factorització QR de la matriu A
Q,R = qr(A,mode='economic')

# 2) Fem el producte de Q^t b:
Qtb = Q.T@b

# 3) Resolem el sistema R x = Qtb
x = solve_triangular(R, Qtb)


############################# RESOLUCIÓ PER QR (v2) ############################

# 1) Aquesta versió de la descomposició ens dona directament Q^T*b
#    a la vegada que calcula R:
Qtb, R = qr_multiply(A, b, mode='right')

# 2) Resolem el sistema triangular:
x = solve_triangular(R, Qtb)
```

* Volem calcular la recta de regressió lineal (la recta que millor aproxima) les següents dades:

```python
import numpy as np
import matplotlib.pyplot as plt

# Generem les dades
m = 500 # Numero de dades
np.random.seed(42)  # Perquè es pugui repoduïr
x = np.random.rand(m) * 10  # Valors de x entre 0 i 10
y = 2 * x + 4 + np.random.randn(m)  # Relació lineal amb soroll

# Visualitzem led dades:
plt.scatter(x, y, label="Dades experimentals",s = 2)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Dades per fer la regressió lineal")
plt.legend()
plt.show()

```

*És a dir volem trobar $c_0$ i $c_1$ tals que $Y = c_1X+c_0$.*

*Notem que per cada valor $(X_i,Y_i)$ amb $i=1\ldots,m$, tenim l'equació:*

$$
\begin{pmatrix}
1 & X_i
\end{pmatrix}
\begin{pmatrix}
c_0 \\ c_1
\end{pmatrix}
= Y_i,
$$

*o, equivalenment, tenim el sistema:*

$$
\begin{pmatrix}
1 & X_1\\
1 & X_2\\
\vdots & \vdots \\
1 & X_{m}
\end{pmatrix}
\begin{pmatrix}
c_0 \\ c_1
\end{pmatrix}
=
\begin{pmatrix}
Y_1\\
Y_2\\
\vdots \\
Y_{m}
\end{pmatrix},
$$

*el qual podem resoldre utilitzant la factorització $QR$.*

```python
from scipy.linalg import solve_triangular, qr_multiply
import numpy as np
import matplotlib.pyplot as plt

# Generem les dades
m = 500 # Numero de dades
np.random.seed(42)  # Perquè es pugui repoduïr
x = np.random.rand(m) * 10  # Valors de x entre 0 i 10
y = 2 * x + 4 + np.random.randn(m)  # Relació lineal amb soroll


# Matriu del sistema:
A = np.empty((m,2))
A[:,0] = 1
A[:,1] = x

# 1) Aquesta versió de la descomposició ens dona directament Q^T*b
#    a la vegada que calcula R:
Qty, R = qr_multiply(A, y, mode='right')

# 2) Resolem el sistema triangular:
c = solve_triangular(R, Qty)


## Postprocess per dibuixar la recta entre 0 i 10:
x_new = np.linspace(0, 10, 100) # 100 valors entre 0 i 10, seria suficient amb 2, ja que és una recta

y_new = c[1]*x_new + c[0] # Solució obtinguda


# Visualitzem led dades + la recta de regressió calculada:
plt.scatter(x, y, label="Dades experimentals", s = 2)
plt.plot(x_new, y_new, label=f"{c[1]}x + {c[0]}",color='black')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Regressió lineal")
plt.legend()
plt.show()
```

## Mètode de Gram-Schmidt

##### Projecció d'un vector en una direcció

Donat un vector $\boldsymbol{v}$ i un vector unitari $\hat{\boldsymbol{u}}$ sempre podem descomposar $\boldsymbol{v}$ en la part paral·lela i perpendicular a $\hat{\boldsymbol{u}}$:

$$
 \boldsymbol{v} = \underbrace{\boldsymbol{v} - \left(\boldsymbol{v}\cdot\hat{\boldsymbol{u}}\right)\hat{\boldsymbol{u}}}_{\text{part perpendicular a }\hat{\boldsymbol{u}}} + \underbrace{\left(\boldsymbol{v}\cdot\hat{\boldsymbol{u}}\right)\hat{\boldsymbol{u}}}_{\text{part paral·lela a }\hat{\boldsymbol{u}}}
$$

i és precissament aquesta propietat el que anirem utilitzant en el procès d'ortogonalització de Gram-Schmidt. Observem que si ho volguesim fer amb un vector $\boldsymbol{u}$ no unitari tindriem:

$$
\begin{aligned}
 \boldsymbol{v} &= \underbrace{\boldsymbol{v} - \left(\boldsymbol{v}\cdot\frac{\boldsymbol{u}}{||\boldsymbol{u}||}\right)\frac{\boldsymbol{u}}{||\boldsymbol{u}||}}_{\text{part perpendicular a }\boldsymbol{u}} + \underbrace{\left(\boldsymbol{v}\cdot\frac{\boldsymbol{u}}{||\boldsymbol{u}||}\right)\frac{\boldsymbol{u}}{||\boldsymbol{u}||}}_{\text{part paral·lela a }\boldsymbol{u}}\\
 & = \underbrace{\boldsymbol{v} - \left(\frac{\boldsymbol{v}\cdot\boldsymbol{u}}{\boldsymbol{u}\cdot\boldsymbol{u}}\right)\boldsymbol{u}}_{\text{part perpendicular a }\boldsymbol{u}} + \underbrace{\left(\frac{\boldsymbol{v}\cdot\boldsymbol{u}}{\boldsymbol{u}\cdot\boldsymbol{u}}\right)\boldsymbol{u}}_{\text{part paral·lela a }\boldsymbol{u}}
 \end{aligned}
$$

##### Ortogonalització d'una base de vectors:

Supososem que tenim $n$ vectors $\boldsymbol{v}_1$, $\boldsymbol{v}_2$, $\ldots$, $\boldsymbol{v}_n$ vectors linealment independents i que volem generar una base $\{\boldsymbol{u}_1,\boldsymbol{u}_2,\ldots,\boldsymbol{u}_n\}$ que generi el mateix subespai que $V = \{\boldsymbol{v}_1,\boldsymbol{v}_2,\ldots,\boldsymbol{v}_n\}$. El que podriem fer per fer-ho el següent procediment iteratiu:

Pel primer element podem simplement fer:

* $ \boldsymbol{u}_1 = \boldsymbol{v}_1$

Pel segon element, triem la part de $\boldsymbol{v}_2$ que és perpendicular a $\boldsymbol{u}_1$:

* $
\displaystyle{
  \boldsymbol{u}_2 = \boldsymbol{v}_2 - \left(\frac{\boldsymbol{v}_2\cdot\boldsymbol{u}_1}{\boldsymbol{u}_1\cdot\boldsymbol{u}_1}\right)\boldsymbol{u}_1}
$

Pel tercer element, triem la part de $\boldsymbol{v}_3$ que és perpendicular a $\boldsymbol{u}_1$ i $\boldsymbol{u}_2$:

* $\displaystyle{\boldsymbol{u}_3=\boldsymbol{v}_3-\left(\frac{\boldsymbol{v}_3\cdot\boldsymbol{u}_1}{\boldsymbol{u}_1\cdot\boldsymbol{u}_1}\right)\boldsymbol{u}_1-\left(\frac{\boldsymbol{v}_3\cdot\boldsymbol{u}_2}{\boldsymbol{u}_2\cdot\boldsymbol{u}_2}\right)\boldsymbol{u}_2}$

I, pel cas general, tenim:

* $\displaystyle{\boldsymbol{u}_k=\boldsymbol{v}_k-\sum_{s=1}^{k-1}\left(\frac{\boldsymbol{v}_k\cdot\boldsymbol{u}_s}{\boldsymbol{u}_s\cdot\boldsymbol{u}_s}\right)\boldsymbol{u}_s}$

&nbsp;

#### Ortonormalització d'una base de vectors

Suposem ara que volem generar una base del mateix subespai que $A = \{\boldsymbol{a}_1,\boldsymbol{a}_2,\ldots,\boldsymbol{a}_n\}$ (on els $\boldsymbol{a}_i$, $i=1,\ldots,n$ són linealment independents) però ara, a més, volem que els vectors siguin ortonormals, anomenarem a aquests vectors $\boldsymbol{q}_1,\ldots,\boldsymbol{q}_n$:

El primer vector, el podem definir com

* $ \displaystyle{\boldsymbol{q}_1 = \frac{\boldsymbol{a}_1}{r_{11}}}$, on $r_{11} = || \boldsymbol{a}_1 ||$,

el segon com:

* $\displaystyle{\boldsymbol{q}_2=\frac{\boldsymbol{a}_2-\color{blue}{\overbrace{\left(\boldsymbol{a}_2\cdot\boldsymbol{q}_1\right)}^{r_{12}}}\boldsymbol{q}_1}{\color{red}{\underbrace{||\boldsymbol{a}_2-\left(\boldsymbol{a}_2\cdot\boldsymbol{q}_1\right)\boldsymbol{q}_1||}_{r_{22}}}}}$,

el tercer com:

* $\displaystyle{\boldsymbol{q}_3=\frac{\boldsymbol{a}_3-\color{blue}{\overbrace{\left(\boldsymbol{a}_3\cdot\boldsymbol{q}_1\right)}^{r_{13}}}\boldsymbol{q}_1-\color{purple}{\overbrace{\left(\boldsymbol{a}_3\cdot\boldsymbol{q}_2\right)}^{r_{23}}}\boldsymbol{q}_2}{\color{red}{\underbrace{||\boldsymbol{a}_3-\left(\boldsymbol{a}_3\cdot\boldsymbol{q}_1\right)\boldsymbol{q}_1-\left(\boldsymbol{a}_3\cdot\boldsymbol{q}_2\right)\boldsymbol{q}_2||}_{r_{33}}}}}$,

i, en general:

* $\displaystyle{\boldsymbol{q}_k=\frac{\boldsymbol{a}_k-\sum_{s=1}^{k-1}\color{blue}{\overbrace{\left(\boldsymbol{a}_k\cdot\boldsymbol{q}_s\right)}^{r_{sk}}}\boldsymbol{q}_s}{\color{red}{\underbrace{||\boldsymbol{a}_k-\sum_{s=1}^{k-1}\left(\boldsymbol{a}_k\cdot\boldsymbol{q}_s\right)\boldsymbol{q}_s||}_{r_{kk}}}}}$,

amb el que hem deduït la factorització de Gram-Schmidt.

&nbsp;

### Mètode d'ortogonalització modificat de Gram-Schmidt

#### Idea de la modificació

Fixeu-vos que és el mateix calcular la part perpendicular (normalitzada) del vector $\boldsymbol{a}_3$ anterior (és a dir, calcular $\boldsymbol{q}_3$) eliminant la part paral·lela a $\boldsymbol{q}_1$ i $\boldsymbol{q}_2$ o bé en dues etapes:

Pimer calculem la part  de $\boldsymbol{a}_3$ perpendicular a $\boldsymbol{q}_1$:

* $\displaystyle{\boldsymbol{v}=\boldsymbol{a}_3-\left(\boldsymbol{a}_3\cdot\boldsymbol{q}_1\right)\boldsymbol{q}_1}$,

i ara, calculem la part de $\boldsymbol{v}$ perpendicular a $\boldsymbol{q}_2$, i la normalitzem per calcular  $\boldsymbol{q}_3$:

* $\displaystyle{\boldsymbol{q}_3=\frac{\boldsymbol{v}-\left(\boldsymbol{v}\cdot\boldsymbol{q}_2\right)\boldsymbol{q}_2}{||\boldsymbol{v}-\left(\boldsymbol{v}\cdot\boldsymbol{q}_2\right)\boldsymbol{q}_2||}}$.


La idea de la factorització modificada és utilitzar això.

#### Algorisme (versió compacta):

Notarem per $q_i$ la i-èssima columna de $Q$.

$
\begin{aligned}
& Q = A\\
& R = 0\\
&\texttt{for } i = 1\div n\\
&\phantom{for } r_{ii} \leftarrow || \boldsymbol{q}_i ||\\
&\phantom{for } \boldsymbol{q}_{i} \leftarrow \boldsymbol{q}_i/r_{ii}\\
&\phantom{for } \texttt{for } j = i+1\div n\\
&\phantom{for for } r_{ij} \leftarrow \boldsymbol{q}_i\cdot \boldsymbol{q}_j\\
&\phantom{for for } \boldsymbol{q}_j \leftarrow \boldsymbol{q}_j - r_{ij} \boldsymbol{q}_i\\
&\phantom{for } \texttt{end for }\\
&\texttt{end for }
\end{aligned}
$

pel cas del sistema associat a $p_2^*$.

A partir de la factorització, ja podriem resoldre el sistema multiplicant el terme independent per $Q^t$ i fent substitució endarrera del sistema resultant.

```python
## Codi per resoldre el problema anterior:

from scipy.linalg import solve_triangular, qr
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(0.25,1.75, 7)
y = np.array([0.4,0.5,0.9,1.28,1.6,1.66,2.02])

# Matriu del sistema:
A = np.empty((len(x),3))
A[:,0] = 1
A[:,1] = x
A[:,2] = x**2

# 1) Calculem la factorització QR:
Q,R = qr(A,mode='economic')

# 2a) Calculem Qt*y per p1 (només utilitzem les dues primeres columens de Q):
Qty_1 = Q[:,:2].T@y

# 3a) Resolem el sistema per substitució endarrera (només utilitzem les dues
#     primers files i columnes de R):
sol1 = solve_triangular(R[:2,:2], Qty_1)

# 2b) Calculem Qt*y per p2:
Qty_2 = Q.T@y

# 3b) Resolem el sistema per substitució endarrera:
sol2 = solve_triangular(R, Qty_2)



## Postprocess per dibuixar les solucions
x_new = np.linspace(0, 2, 100) # 100 valors entre 0 i 2

y_1 = sol1[1]*x_new + sol1[0] # p1
y_2 = sol2[2]*x_new**2 + sol2[1]*x_new + sol2[0] # p2


# Visualitzem led dades + la recta de regressió calculada:
plt.scatter(x, y, label="Dades experimentals", s = 2)
plt.plot(x_new, y_1, label=f"{sol1[1]:6f}x{sol1[0]:+6f}",color='blue')
plt.plot(x_new, y_2, label=f"{sol2[2]:6f}x^2{sol2[1]:+6f}x{sol2[0]:+6f}",color='red')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Regressió lineal")
plt.legend()
plt.show()
```

```python
## Calculem, per últim, l'error (a partir de l'apartat anterior de codi):


p_1 = sol1[1]*x + sol1[0] # p1
p_2 = sol2[2]*x**2 + sol2[1]*x+ sol2[0] # p2

dif1 = y-p_1
dif2 = y-p_2

norma1 = np.dot(dif1,dif1)
norma2 = np.dot(dif2,dif2)

print("Per p1, tenim:",norma1)
print("Per p2, tenim:",norma2)

```
