# Factorització $QR$ mitjançant transformacions de Householder

## Matrius de Householder

Per cada vector no nul $\boldsymbol{u}\in\mathbb{R}^n$ definim la seva matriu de Householder associada com:

$$
    P(\boldsymbol{u}) := Id - \frac{2}{\boldsymbol{u}^\top \boldsymbol{u}} \boldsymbol{u} \boldsymbol{u}^\top.
$$

Fixem-nos que $\boldsymbol{u}$ és un vector columna i, per tant,


*   $\boldsymbol{u}^\top \boldsymbol{u} = \boldsymbol{u}\cdot \boldsymbol{u} = \|\boldsymbol{u}\|^2$ és un escalar.

*   $\boldsymbol{u} \boldsymbol{u}^\top$ és una matriu simètrica:

$$
\boldsymbol{u} \boldsymbol{u}^\top =
\begin{pmatrix}
u_1^2 & u_1 u_2 & \cdots & u_1 u_n\\
u_1 u_2 & u_2^2 & \cdots & u_2 u_n\\
\vdots & \vdots & \ddots & \vdots\\
u_1 u_n & u_2 u_n & \cdots & u_n^2\\
\end{pmatrix}.
$$



```python
import numpy as np

# Matriu Householder:
def matHouseholder(u):
    n = len(u)
    H = np.eye(n) - 2*np.outer(u,u)/np.dot(u,u)
    return H

# Declarem el vector:
u = np.array([1,2,3,4],dtype = np.float64)

# Construim la matriu:
H = matHouseholder(u)

# Printem la matriu:
print(H)
```

### Propietats bàsiques de les matrius de Householder:

Sigui $\boldsymbol{u}\in\mathbb{R}^n\setminus\{\boldsymbol{0}\}$. Aleshores:

1.   $P(k \boldsymbol{u}) = P(\boldsymbol{u}), \forall k \neq 0$.

2.  $P(\boldsymbol{u})\boldsymbol{u} = - \boldsymbol{u}$

2.   $P(\boldsymbol{u})$ és una matriu simètrica ($P(\boldsymbol{u}) = P(\boldsymbol{u})^\top$).

3.   $P(\boldsymbol{u})P(\boldsymbol{u}) = Id$

4.   $P(\boldsymbol{u}) = P(\boldsymbol{u})^{-1} = P(\boldsymbol{u})^\top$, és a dir, $P(\boldsymbol{u})$ és una matriu ortonormal.

5.   $\det(P(\boldsymbol{u})) = -1$.
  

##### *Demostració*

1. Veiem-ho a partir de la definició:

$$
P(k\boldsymbol{u}) : = Id - \frac{2}{(k\boldsymbol{u})^\top (k\boldsymbol{u})} (k\boldsymbol{u}) (k\boldsymbol{u}^\top) = Id - \frac{2}{k^2\boldsymbol{u}^\top \boldsymbol{u}} k^2\boldsymbol{u} \boldsymbol{u}^\top
= Id - \frac{2}{\boldsymbol{u}^\top \boldsymbol{u}} \boldsymbol{u} \boldsymbol{u}^\top =: P(\boldsymbol{u})
$$

&nbsp;

2. Ho tornarem a veure a partir de la definició:

$$
P(\boldsymbol{u})\boldsymbol{u} = \left(Id - \frac{2}{\boldsymbol{u}^\top \boldsymbol{u}} \boldsymbol{u} \boldsymbol{u}^\top\right) \boldsymbol{u} = \boldsymbol{u} - \frac{2}{\boldsymbol{u}^\top \boldsymbol{u}} (\boldsymbol{u} \boldsymbol{u}^\top) \boldsymbol{u}
= \boldsymbol{u} - \frac{2}{\boldsymbol{u}^\top \boldsymbol{u}} \boldsymbol{u} (\boldsymbol{u}^\top \boldsymbol{u})
= \boldsymbol{u} - 2\boldsymbol{u} = - \boldsymbol{u}
$$

&nbsp;

3. Trivial, la suma de dues matrius simètriques és simètrica.

&nbsp;

4. Veiem-ho a partir de la definició:

$$
\begin{aligned}
    P(\boldsymbol{u})P(\boldsymbol{u}) &:= \left(Id - \frac{2}{\boldsymbol{u}^\top \boldsymbol{u}} \boldsymbol{u} \boldsymbol{u}^\top\right)
    \left(Id - \frac{2}{\boldsymbol{u}^\top \boldsymbol{u}} \boldsymbol{u} \boldsymbol{u}^\top\right)\\
    & = Id - 2 \frac{2}{\boldsymbol{u}^\top \boldsymbol{u}} \boldsymbol{u} \boldsymbol{u}^\top + \left(\frac{2}{\boldsymbol{u}^\top \boldsymbol{u}} \boldsymbol{u} \boldsymbol{u}^\top\right)\left(\frac{2}{\boldsymbol{u}^\top \boldsymbol{u}} \boldsymbol{u} \boldsymbol{u}^\top\right)\\
    & = Id - \frac{4}{\boldsymbol{u}^\top \boldsymbol{u}} \boldsymbol{u} \boldsymbol{u}^\top
    + \frac{4}{(\boldsymbol{u}^\top \boldsymbol{u}) (\boldsymbol{u}^\top \boldsymbol{u})} (\boldsymbol{u} \boldsymbol{u}^\top)(\boldsymbol{u} \boldsymbol{u}^\top)\\
    & = Id - \frac{4}{\boldsymbol{u}^\top \boldsymbol{u}} \boldsymbol{u} \boldsymbol{u}^\top
    + \frac{4}{(\boldsymbol{u}^\top \boldsymbol{u}) (\boldsymbol{u}^\top \boldsymbol{u})} \boldsymbol{u} (\boldsymbol{u}^\top\boldsymbol{u}) \boldsymbol{u}^\top\\
    & = Id - \frac{4}{\boldsymbol{u}^\top \boldsymbol{u}} \boldsymbol{u} \boldsymbol{u}^\top
    + \frac{4}{\boldsymbol{u}^\top \boldsymbol{u}} \boldsymbol{u} \boldsymbol{u}^\top\\
    & = Id
\end{aligned}
$$


&nbsp;

5. Conseqüència de 3 i 4.

&nbsp;

6. Per simplificar els càlculs definim el vector unitari
$
\hat{\boldsymbol{u}} = \cfrac{\boldsymbol{u}}{\|\boldsymbol{u}\|}
$, així, per 1, tenim:

$$ P(\hat{\boldsymbol{u}})  =  P(\boldsymbol{u}) $$

i es pot veure que es satisfà la següent igualtat:

$$
\begin{aligned}
\begin{pmatrix}
Id & \boldsymbol{0}\\
-\hat{\boldsymbol{u}}^\top & 1
\end{pmatrix}
\begin{pmatrix}
P(\hat{\boldsymbol{u}}) & 2\hat{\boldsymbol{u}}\\
\boldsymbol{0}^\top & 1
\end{pmatrix}
\begin{pmatrix}
Id & \boldsymbol{0}\\
\hat{\boldsymbol{u}}^\top & 1
\end{pmatrix}
& =
\begin{pmatrix}
P(\hat{\boldsymbol{u}}) & 2\hat{\boldsymbol{u}}\\
\hat{\boldsymbol{u}}^\top & -1
\end{pmatrix}
\begin{pmatrix}
Id & \boldsymbol{0}\\
\hat{\boldsymbol{u}}^\top & 1
\end{pmatrix}\\
& =
\begin{pmatrix}
P(\hat{\boldsymbol{u}}) +2\hat{\boldsymbol{u}}\hat{\boldsymbol{u}}^\top & 2\hat{\boldsymbol{u}}\\
\boldsymbol{0}^\top & -1
\end{pmatrix}\\
& =
\begin{pmatrix}
Id & 2\hat{\boldsymbol{u}}\\
\boldsymbol{0}^\top & -1
\end{pmatrix}
\end{aligned}
$$

Si prenem determinant a banda i banda, tenim:

$$
\underbrace{\det\begin{pmatrix}
Id & \boldsymbol{0}\\
-\hat{\boldsymbol{u}}^\top & 1
\end{pmatrix}}_{ = 1}
\underbrace{
\det\begin{pmatrix}
P(\hat{\boldsymbol{u}}) & 2\hat{\boldsymbol{u}}\\
\boldsymbol{0}^\top & 1
\end{pmatrix}}_{= \det(P(\hat{\boldsymbol{u}}))}
\underbrace{
\det\begin{pmatrix}
Id & \boldsymbol{0}\\
\hat{\boldsymbol{u}}^\top & 1
\end{pmatrix}}_{ = 1}
=
\underbrace{
\det\begin{pmatrix}
Id & 2\hat{\boldsymbol{u}}\\
\boldsymbol{0}^\top & -1
\end{pmatrix}}_{ = -1}.
$$

Així,
$$\det(P(\boldsymbol{u})) = \det(P(\hat{\boldsymbol{u}})) = -1.$$

### Interpretació geomètrica de les Matrius de Householder:



L'aplicació $\boldsymbol{x}\in\mathbb{R}^n \mapsto P(\boldsymbol{u})\boldsymbol{x}\in\mathbb{R}^n $ és una **simetria** (o **reflexió**) respecte l'hiperplà perpendicular a $\boldsymbol{u}$.

&nbsp;

Recordem que podem descomposar $\boldsymbol{x}$ en la seva part paral·lela i perpendicular a $\boldsymbol{u}$ de la següen forma:

$$
\begin{aligned}
\boldsymbol{x} &= \underbrace{\color{blue}{\boldsymbol{u} - \left(\frac{\boldsymbol{u}^\top\boldsymbol{x}}{\boldsymbol{u}^\top\boldsymbol{u}}\right)\boldsymbol{u}}}_{\text{part perpendicular a }\boldsymbol{u}} + \underbrace{\color{purple}{\left(\frac{\boldsymbol{u}^\top\boldsymbol{x}}{\boldsymbol{u}^\top\boldsymbol{u}}\right)}\boldsymbol{u}}_{\text{part paral·lela a }\boldsymbol{u}}\\
& = \color{blue}{\boldsymbol{v}} + \color{purple}{c}\boldsymbol{u}
\end{aligned}
$$

Notem que $\boldsymbol{v}$ és la part de perpendicular de $\boldsymbol{x}$ a $\boldsymbol{u}$ i, per tant, $\boldsymbol{u}^\top\boldsymbol{v} = 0$. Calculem ara el resultat de la transformació:

$$
\begin{aligned}
P(\boldsymbol{u})\boldsymbol{x}
&= P(\boldsymbol{u}) (\boldsymbol{v} + c\boldsymbol{u})\\
&= P(\boldsymbol{u})\boldsymbol{v} + cP(\boldsymbol{u})\boldsymbol{u}\\
& = \left(Id - \frac{2}{\boldsymbol{u}^\top \boldsymbol{u}} \boldsymbol{u} \boldsymbol{u}^\top\right)\boldsymbol{v} + c\left(Id - \frac{2}{\boldsymbol{u}^\top \boldsymbol{u}} \boldsymbol{u} \boldsymbol{u}^\top\right)\boldsymbol{u}\\
& = \boldsymbol{v} - \frac{2}{\boldsymbol{u}^\top \boldsymbol{u}} (\boldsymbol{u} \boldsymbol{u}^\top)\boldsymbol{v} + c\left(\boldsymbol{u} - \frac{2}{\boldsymbol{u}^\top \boldsymbol{u}} (\boldsymbol{u} \boldsymbol{u}^\top)\boldsymbol{u}\right)\\
& = \boldsymbol{v} - \frac{2}{\boldsymbol{u}^\top \boldsymbol{u}} \boldsymbol{u} (\underbrace{\boldsymbol{u}^\top\boldsymbol{v}}_{=0}) + c\left(\boldsymbol{u} - \frac{2}{\boldsymbol{u}^\top \boldsymbol{u}} \boldsymbol{u} (\boldsymbol{u}^\top\boldsymbol{u})\right)\\
& = \boldsymbol{v} + c(\boldsymbol{u}- 2\boldsymbol{u})\\
& = \boldsymbol{v} - c\boldsymbol{u}.
\end{aligned}
$$

### Transformació entre dos vectors amb el mateix mòdul via matrius de Householder

Suposem que tenim els vector $\boldsymbol{a},\,\boldsymbol{b}\in\mathbb{R}^n\setminus\{\boldsymbol{0}\}$ tal que $\|\boldsymbol{a}\| = \|\boldsymbol{b}\|$.

**Objectiu:** Trobar $\boldsymbol{u}$ tal que $\boldsymbol{a} = P(\boldsymbol{u})\boldsymbol{b}$ (i, per tant, $\boldsymbol{b} = P(\boldsymbol{u})\boldsymbol{a}$).


Suposem que $\boldsymbol{a} = \boldsymbol{v} + c\boldsymbol{u}$. Per l'apartat anterior sabem que $\boldsymbol{b} = \boldsymbol{v} - c\boldsymbol{u}$ i, per tant, tindriem:

$$
    \boldsymbol{a} - \boldsymbol{b} = 2c\boldsymbol{u}.
$$

Així doncs, per la primera propietat bàsica de les matrius de Householder, podem triar:

$$
    \boldsymbol{u} = \boldsymbol{a} - \boldsymbol{b}.
$$

### Factorització $QR$ via matrius de Householder

#### Idea:

Podem utilitzar el resultat anterior per fer calcular la factorització  $QR$. És a dir, anirem utilitzant les matrius de Householder per a construir la factorització. En particular, si tenim la matriu $A\in\mathcal{M}_{m\times n}(\mathbb{R})$ ($m \geq n$) de rang màxim, tindrem:

$$
\tilde{Q} = \left(\begin{array}{ccc|ccc} \\& Q & & \boldsymbol{q}_{n+1} & \cdots & \boldsymbol{q}_m \\ &\end{array}\right),\qquad \tilde{R} = \left(\begin{array}{ccc} \\ & R & \\ \\\hline 0 & \cdots & 0\\ \vdots & & \vdots\\ 0 & \cdots & 0\end{array}\right),
$$

on $\tilde{Q}\in \mathcal{M}_m(\mathbb{R})$ i $\tilde{R} \in \mathcal{M}_{m\times n}(\mathbb{R})$. Observem que:

$$
    A = Q R = \tilde{Q}\tilde{R}.
$$

Així, tindrem:

* $\tilde{Q} = \tilde{H}_1\cdot\ldots\cdot \tilde{H}_n$

* $\tilde{R} = \tilde{H}_n\cdot\ldots\cdot \tilde{H}_1 A$

on les $\tilde{H}_i$, amb $i=1,\ldots,n$, són matrius de Householder.

&nbsp;

Així, partint de la matriu $A=A^{(1)}$, mitjançant les matrius de Householder anirem triangularitzant la matriu, és a dir:

$$
A^{(1)} =
\begin{pmatrix}
a_{11}^{(1)} & a_{12}^{(1)} & \cdots & a_{1n}^{(1)}\\
a_{21}^{(1)} & a_{22}^{(1)} & \cdots & a_{2n}^{(1)}\\
\vdots & \vdots & & \vdots\\
a_{m1}^{(1)} & a_{m2}^{(1)} & \cdots & a_{mn}^{(1)}\\
\end{pmatrix},
$$

$$
A^{(2)} = H_1 A^{(1)} =
\begin{pmatrix}
a_{11}^{(2)} & a_{12}^{(2)} & \cdots & a_{1n}^{(2)}\\
0 & a_{22}^{(2)} & \cdots & a_{2n}^{(2)}\\
\vdots & \vdots & & \vdots\\
0 & a_{m2}^{(2)} & \cdots & a_{mn}^{(2)}\\
\end{pmatrix},
$$


$$
A^{(k)} = H_{k-1} A^{(k-1)} = H_{k-1}\ldots H_1 A^{(1)} =
\begin{pmatrix}
a_{1,1}^{(2)} & a_{1,2}^{(2)} & \cdots & a_{1,k-1}^{(2)} & a_{1,k}^{(2)} & \cdots & a_{1,n}^{(2)} \\
              & a_{2,2}^{(3)} & \cdots & a_{1,k-1}^{(2)} & a_{2,k}^{(3)} & \cdots & a_{2n}^{(3)} \\
              &               &  \ddots & \vdots &\vdots      &        & \vdots       \\
              &               &        & a_{k-1,k-1}^{(k)} & a_{k-1,k}^{(k)} & \cdots & a_{k-1,n}^{(k)} \\
              &               &        &        & a_{k,k}^{(k)} & \cdots & a_{k,n}^{(k)} \\
              &               &        &        & \vdots       &        & \vdots       \\
              &               &        &        & a_{m,k}^{(k)} & \cdots & a_{m,n}^{(k)} \\
\end{pmatrix},
$$

$$
A^{(n+1)} = H_{n} A^{(n)} = H_{n}\ldots H_1 A^{(1)} =
\begin{pmatrix}
a_{1,1}^{(2)} & a_{1,2}^{(2)} & \cdots &  a_{1,n}^{(2)} \\
              & a_{2,2}^{(3)} & \cdots &  a_{2,n}^{(3)} \\
              &               & \ddots & \vdots       \\
              &               &        & a_{n,n}^{(n+1)} \\ \hline
 \\
 \\
 \\
\end{pmatrix}.
$$

#### Primera etapa de l'algorisme

Si denotem per $\boldsymbol{a}_1^{(1)}$ la primera columna de $A^{(1)}$, ens agradia enviar aquest vector a un vector del tipus $\boldsymbol{a}_1^{(2)} = (a_{11}^{(2)}, 0, \ldots, 0)^\top$.

Pel que hem vist abans, podriem fer-ho via la matriu de Householder $H_1 = P(\boldsymbol{a}_1^{(1)}-\boldsymbol{a}_1^{(2)})$ però necessitem que els mòduls de $\boldsymbol{a}_1^{(1)}$ i $\boldsymbol{a}_1^{(2)}$ siguin iguals.

Per tant, podem triar $a_{11}^{(2)} = \pm \|\boldsymbol{a}_1^{(1)} \|$.

Normalment, per tal d'evitar errors amb possibles cancel·lacions (restes de quantitats similars) el que farem es triar, $a_{11}^{(2)}$ amb el signe oposat a $a_{11}^{(1)}$:

$$
a_{11}^{(2)} = -\text{sign}(a_{11}^{(1)}) \|\boldsymbol{a}_1^{(1)} \|.
$$

D'aquesta manera, considerant el vector canònic $\boldsymbol{e}_1 = (1,0,\ldots,0)^\top$, tenim:

$$
    H_1 = P(\boldsymbol{a}_1^{(1)} + \text{sign}(a_{11}^{(1)}) \|\boldsymbol{a}_1^{(1)} \| \boldsymbol{e}_1).
$$

Així, podriem obtenir fàcilment $A^{(2)}$:

$$
A^{(2)} = H_1 A^{(1)}
$$

&nbsp;

##### **Cost computacional:**

Anem a calcular el cost computacional d'aquest càlcul:

Càlcul de $\boldsymbol{u} = \boldsymbol{a}_1^{(1)} + \text{sign}(a_{11}^{(1)}) \|\boldsymbol{a}_1^{(1)} \| \boldsymbol{e}_1$:

* Càlcul $\|\boldsymbol{a}_1^{(1)} \|$:
 * $m$ productes
 * $m-1$ sumes
 * $1$ arrel

* $u_1 = a_{11}^{(1)} + \text{sign}(a_{11}^{(1)})\|\boldsymbol{a}_1^{(1)} \|$
 * $1$ suma
 * Canvi de bit de signe ("menyspreable")

Així, el nombre total d'operacions seria $2m+1$.

&nbsp;




Enlloc de construir i multiplicar la matriu fixem-nos que podem agrupar els productes escalars de manera astuta per evitar-nos operacions.

Primer de tot, calculem el terme:
$$
\beta = \frac{2}{\boldsymbol{u}^\top \boldsymbol{u}}
$$


A continuació, calculem cada vector $\boldsymbol{a}_{j}^{(2)}$ per $j >1$

$$
\boldsymbol{a}_{j}^{(2)} = P(\boldsymbol{u})\boldsymbol{a}_{j}^{(1)}
 = (Id - \beta \boldsymbol{u}\boldsymbol{u}^\top)\boldsymbol{a}_{j}^{(1)}
 = \boldsymbol{a}_{j}^{(1)} - \boldsymbol{u}(\beta(\boldsymbol{u}^\top\boldsymbol{a}_{j}^{(1)}))
$$

el qual té el següent cost:

* $\boldsymbol{u}^\top\boldsymbol{a}_{j}^{(1)}$:
 * $m$ multiplicacions
 * $m-1$ sumes
* $\beta(\boldsymbol{u}^\top\boldsymbol{a}_{j}^{(1)})$:
 * $1$ multiplicació:
* $\boldsymbol{u}(\beta(\boldsymbol{u}^\top\boldsymbol{a}_{j}^{(1)}))$:
 * $m$ multiplicacions
* $\boldsymbol{a}_{j}^{(1)} - \boldsymbol{u}(\beta(\boldsymbol{u}^\top\boldsymbol{a}_{j}^{(1)}))$:
 * $m$ restes

&nbsp;

Per tant, podem reduir el cost $O(m^2n)$ de multiplicar les matrius a un cost $O(mn)$.
