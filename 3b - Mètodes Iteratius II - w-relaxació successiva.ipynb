## Mètode de $\omega$-relaxació successiva

Sigui $A\in \mathcal{M}_n(\mathbb{R})$ una matriu de rang màxim i sigui $\boldsymbol{b}$ un vector de $\mathbb{R}^n$. Definim les matrius $L$, $D$ i $U$ de la següen manera:

* $\displaystyle{L=\begin{pmatrix}0 &  &  &  & \\a_{2,1} &0 & & & \\ \vdots & \ddots & \ddots & \\ \vdots & & \ddots & 0 & \\ a_{n,1} & \cdots & \cdots & a_{n,n-1}& 0\end{pmatrix}}$

* $\displaystyle{D=\begin{pmatrix}a_{1,1} \\ & a_{2,2}\\ & & \ddots\\ & & & a_{n,n}\end{pmatrix}}$

* $\displaystyle{U=\begin{pmatrix}0 & a_{1,2} & \cdots & \cdots & a_{1,n}\\ &0 &\ddots & & \vdots \\& & \ddots & \ddots & \vdots\\ & & & 0 & a_{n-1,n}\\ & & & & 0\end{pmatrix}}$

Podem construir el **mètode de $\omega$-relaxació successiva** de la següent forma:

$$
\begin{aligned}
A\boldsymbol{x} = (L+D+U)\boldsymbol{x} = \boldsymbol{b} &\Longleftrightarrow \omega(L+D+U)\boldsymbol{x} = \omega\boldsymbol{b}\\
&\Longleftrightarrow \omega L\boldsymbol{x} = -\omega U\boldsymbol{x} - \omega D\boldsymbol{x} + \omega\boldsymbol{b}\\
&\Longleftrightarrow {\color{blue}{D\boldsymbol{x}} +} \omega L\boldsymbol{x} = {\color{blue}{D\boldsymbol{x}}}-\omega U\boldsymbol{x} - \omega D\boldsymbol{x} + \omega\boldsymbol{b}
\end{aligned}
$$

on $\omega\neq 0$. Agrupem ara termes a banda i banda, i tenim:

$$
\begin{aligned}
 (D+\omega L)\boldsymbol{x} &= D\boldsymbol{x}-\omega U\boldsymbol{x} - \omega D\boldsymbol{x} + \omega\boldsymbol{b}\\
 & = -\omega U\boldsymbol{x} + (1-\omega)D\boldsymbol{x} + \omega\boldsymbol{b}\\
 & = -\big(\omega U + (\omega-1)D\big)\boldsymbol{x} + \omega\boldsymbol{b}
 \end{aligned}
$$

El qual, és equivalent a:

$$
\boldsymbol{x} = -(D+\omega L)^{-1}\big(\omega U + (\omega-1)D\big)\boldsymbol{x} + \omega(D+\omega L)^{-1}\boldsymbol{b}
$$

D'aquesta forma, el **mètode de $\omega$-relaxació successiva** ve donat per:

$$
\boldsymbol{x}^{(k+1)} = \underbrace{-(D+\omega L)^{-1}\big(\omega U + (\omega-1)D\big)}_{B_{\omega}}\boldsymbol{x}^{(k)} + \underbrace{\omega(D+\omega L)^{-1}\boldsymbol{b}}_{\boldsymbol{c}_{\omega}}.
$$



&nbsp;

Fixem-nos que en la pràctica no calcularem la inversa de $(D+\omega L)$ i, per tant, tindrem:

$$
(D+\omega L)\boldsymbol{x}^{(k+1)} = -\big(\omega U + (\omega-1)D\big)\boldsymbol{x}^{(k)} + \omega\boldsymbol{b}
$$

El qual ens dona un sistema triangular inferior (ja que $(D+\omega L)$ és una matriu triangular inferior) que podem resoldre fàcilment per substitució endarrere, obtenint així la descripció del mètode component a component:

* $\displaystyle{x_1^{(k+1)} = (1-\omega) x_1^{(k)} + \frac{\omega}{a_{11}} \left(b_1 - \sum_{j=2}^n a_{1j}x_j^{(k)}\right)}$

* $\displaystyle{x_i^{(k+1)} = (1-\omega) x_i^{(k)} + \frac{\omega}{a_{ii}} \left(b_i - \sum_{j=1}^{i-1}a_{ij}x_j^{(k+1)}- \sum_{j=i+1}^n a_{ij}x_j^{(k)}\right)},\quad i = 2,\ldots,n-1$

* $\displaystyle{x_n^{(k+1)} = (1-\omega) x_n^{(k)} + \frac{\omega}{a_{nn}} \left(b_i - \sum_{j=1}^{n-1}a_{nj}x_j^{(k+1)} \right)}$

&nbsp;


> ⚠️**Observació:**
>
>Fixem-nos que prenent $\omega = 1$ recuperem el mètode de Gauss-Seidel.

&nbsp;

> ⚠️ **Observació II:**
>
> El mètode de $\omega$-relaxació és un mètode de descomposició, però haurem de dividir per $\omega$ la matriu anterior:
>
>$$A = \underbrace{\left(\frac{1}{\omega} D + L\right)}_{M_\omega} - \underbrace{\left(\frac{1-\omega}{\omega}D - U\right)}_{N_\omega}$$
>
>A la pràctica podem utilitzar la matriu del mètode anterior o bé la definida per $M^{-1}_\omega N_\omega$. (El mateix passa per al terme independent del mètode).


### Condició necessària per a la convergència del mètode

> ##### **Proposició:**
>
> Una condició necessària per a la convergència del mètode anterior és que $\omega\in(0,2)$.

#### *Exercici 8*

*Per a resoldre un sistema lineal $A \boldsymbol{x} =\boldsymbol{b}$ amb $A\in M_{n}(\mathbb{C})$ es considera el següent mètode de relaxació: donat
$\boldsymbol{x}^{(0)}=\left(x_1^{(0)}, x_2^{(0)}, \ldots, x_n^{(0)}\right)^{\top}$, per $k\geq 0$ definim:*

$$
\left\{
\begin{aligned}
r_i^{(k)} &= b_i - \sum_{j=1}^{i-1} a_{ij} x_j^{(k+1)} -
\sum_{j=i+1}^n a_{ij} x_j^{(k)}, \\
x_i^{(k+1)} &= (1-\omega) x_i^{(k)} + \omega \ \frac{r_i^{(k)}}{a_{ii}},
\end{aligned}\right.
$$

*per $i=1\div n$ i amb $\omega\in \mathbb{R}$ un paràmetre. Abusant de notació, si $i=1$ prenem $\sum\limits_{j=1}^{i-1} a_{ij} x_j^{(k+1)} \equiv
0$ a la primera fòrmula.*


$a)$ *Trobeu la forma explícita de la matriu d'iteració corresponent.*

$b)$ *Demostreu que si aquest mètode convergeix aleshores $0<\omega<2$.*

$c)$ *Comproveu que si prenem $\omega=1$ aquest mètode es redueix al mètode de Gauss-Seidel. En el cas que $1<\omega<2$ aquest mètode s'anomena de **sobre-relaxació successiva** (en anglès, $\textsf{SOR}$, de successive over-relaxation).*

&nbsp;



###### *Resolució*:

Apartats $a)$ i $c)$ fets.

Recordem l'apartat $b)$:

$$
\begin{aligned}
\prod_{i=1}^n \lambda_i &= \det (B_\omega)\\
&= \det\Big(-(D+\omega L)^{-1}\big(\omega U + (\omega-1)D\big)\Big)\\
&= - \det\left((D+\omega L)^{-1}\right)\,\det\big(\omega U + (\omega-1)D\big)\\
& = -\frac{\det\big(\omega U + (\omega-1)D\big)}{\det\left(D+\omega L\right)}\\
& = - \frac{\prod_{i=1}^n (w-1)a_{ii}}{\prod_{i=1}^n a_{ii}}\\
& = - \frac{(w-1)^n\prod_{i=1}^n a_{ii}}{\prod_{i=1}^n a_{ii}}\\
& = - (w-1)^n
\end{aligned}
$$

Així doncs, tenim:

$$
    \rho(B_\omega) = \max_{i=1,\ldots,n} |\lambda_i| \geq \sqrt[n]{|\omega-1|^n} = |\omega-1|.
$$

Com perquè el mètode sigui convergent necessitem $1>\rho(B_\omega) \geq |\omega-1|$.

#### *Exercici:*

*Implementa una funció que, donada la matriu $A$ i el vector $\boldsymbol{b}$ del sistema $A\boldsymbol{x} = \boldsymbol{b}$, **un vector inicial** $\boldsymbol{x}^{(0)}$ i un valor $\omega$ calculi la solució d'aquest sistema emprant el mètode de $\omega$-relaxació successiva. **No podeu calcular la inversa de la matriu** $(D+\omega L)$.*

*Utilitzeu els tres paràmetres opcionals següents:*

* *`abstol=1.e-10`, que ens indiqui la tolerància mínima amb la qual volem la solució utilitzant la norma euclídea: $\|\boldsymbol{x}^{(k+1)}-\boldsymbol{x}^{(k)}\| < \varepsilon$. Un cop assolida, el mètode ha de retornar la solució.*

* *`reltol=1.e-10`, que ens indiqui la tolerància mínima amb la qual volem la solució utilitzant la norma euclídea: $\|\boldsymbol{x}^{(k+1)}-\boldsymbol{x}^{(k)}\|/\|\boldsymbol{x}^{(k+1)}\| < \varepsilon$. Un cop assolida, el mètode ha de retornar la solució.*

* *`maxIter=100`, que controli el nombre màxim d'iteracions possibles.*

*La funció ha de retornar la solució `x` obtinguda i el nombre d'iteracions realitzades `iter`, en aquest ordre. Si no s'ha assolit la tolerància absoluta `iter` serà -1, si no s'ha assolit la tolerància relativa -2, i si no s'han asolit cap de les dues -3.*

```python
#### Espai per fer l'exercici
import numpy as np
from scipy.linalg import solve_triangular


def sobreRelax(A, b, x, w, abstol=1.e-10, reltol=1.e-10, maxIter=100):

    L = np.tril(A, -1)
    D = np.diag(np.diag(A))
    U = np.triu(A, 1)

    xold = x.copy()

    for iter in range(maxIter):

        Uxb = -(w*U + (w - 1)*D)@xold + w*b

        x = solve_triangular(D + w*L, Uxb, lower = True)

        absolut = np.linalg.norm(x - xold)
        relatiu = absolut / np.linalg.norm(x)

        if (absolut < abstol and relatiu < reltol):
            return x, iter + 1

        xold = x.copy()

    if (absolut >= abstol and relatiu >= reltol):
        return x, -3

    elif (relatiu >= reltol):
        return x, -2

    else:
        return x, -1

A = np.array([[16,3],
              [7,-11]], dtype=np.float64)

b = np.array([11, 13], dtype=np.float64)
xold = np.array([1, 1], dtype=np.float64)


x, iter = sobreRelax(A, b, xold, 0.9)

print(x)
print(iter)
```

### Teorema de Gerschgorin

> #### **Definició:** Discs de Gerschgorin
>
>Sigui $A\in\mathcal{M}_n(\mathbb{C})$ una matriu. Per cada fila $i=1,\ldots,n$ de la matriu considerem:
>$$R_i = \sum_{j\neq i} |a_{ij}|,$$
>i, definim els **discs de Gerschgorin per files** com
>$$D_i = \{z\in\mathbb{C}\,\,:\,\,|z-a_{ii}| \leq R_i\} = \bar{D}(a_{ii},R_i),\qquad \forall i = 1,\ldots, n.$$
>
>&nbsp;
>
>De la mateixa manera, per cada columna $j=1,\ldots,n$ de la matriu, considerem:
>$$\mathcal{R}_j = \sum_{i\neq j} |a_{ij}|,$$
>i, definim els **discs de Gerschgorin per columnes** com
>$$\mathcal{D}_j = \{z\in\mathbb{C}\,\,:\,\,|z-a_{jj}| \leq \mathcal{R}_j\} = \bar{D}(a_{jj},\mathcal{R}_j),\qquad \forall j = 1,\ldots, n.$$



>#### **Teorema de Gerschgorin**
>
>Els valors propis de la matriu $A$ es troben als discs per files de Gerschgorin, és a dir per tot valor propi $\lambda$ de la matriu $A$, tenim que $\lambda\in D_i$ per algun $i=1,\ldots,n$.

###### *Demostració:*

(pissarra)

> ##### **Corol·lari:**
>
> Els valors propis de la matriu $A$ es troben als discs per columnes de Gerschgorin, és a dir per tot valor propi $\lambda$ de la matriu $A$, tenim que $\lambda\in \mathcal{D}_j$ per algun $j=1,\ldots,n$.

&nbsp;

> ##### **Corol·lari 2:**
>
>Per tot valor propi $\lambda$ de $A$ és té
>$$\lambda \in \left(\bigcup_{i=1}^n D_i\right) \cap \left(\bigcup_{j=1}^n \mathcal{D}_j\right).$$

#### *Exemple 1*:

$$
A = \begin{pmatrix} 6 & 0 & 0 & -1\\ 1 & 2 & 0 & 2\\ -1 & 0 & -6 & 1\\-1 & 1 & 0 & -3 \end{pmatrix}
$$

```python
## Discs de Gerschgorin per files:

import numpy as np
import matplotlib.pyplot as plt

# Aquesta funció ens retorna els centres i radis dels discos
# de Gerschgorin per files.
def discsGerschgorin(A):
    # Obtenim dimensions de la matriu A:
    n, m = A.shape

    # Si la matriu no es quadrada retornem error:
    if n != m:
        raise Exception("Error! La matriu no és quadrada")

    # Calculem els centres:
    c = np.diag(A)

    # Calculem els radis:
    #r = np.sum(np.abs(A),axis=1) - np.abs(c)
    r = np.linalg.norm(A, axis =1, ord=1) - np.abs(c)

    return c, r


## Procediment general:

# Definim la matriu:
A = np.array([[6,0,0,-1] ,[1,2,0,2] ,[-1,0,-6,1] ,[-1,1,0,-3]])

# Calculem els vaps i els veps:
vaps, veps = np.linalg.eig(A)
c,r = discsGerschgorin(A)

#n = len(A)

## Dibuixem:
fig, ax = plt.subplots()

ax.set_aspect('equal')  # Fem l'escala dels eixos iguals
ax.grid(True) # Afegim el grid

# Anem dibuixant cada disc:
for i in range(len(A)):
    circle = plt.Circle((np.real(c[i]), np.imag(c[i])), r[i], color='blue', fill=True, alpha=0.5)
    ax.add_patch(circle)

# Dibuixem els VAPS:
plt.scatter(np.real(vaps),np.imag(vaps),color="black", s=5)

# Limits dels eixos:
plt.xlim(-10, 10)
plt.ylim(-10, 10)

# Labels dels eixos:
plt.xlabel('Part real')
plt.ylabel('Part imaginària')

plt.show()
```

```python
## Fem el mateix per columnes:
c2,r2 = discsGerschgorin(A.T)


## Dibuixem:
fig, ax = plt.subplots()

ax.set_aspect('equal')  # Fem l'escala dels eixos iguals
ax.grid(True) # Afegim el grid

# Anem dibuixant cada disc:
for i in range(len(A)):
    circle = plt.Circle((np.real(c2[i]), np.imag(c2[i])), r2[i], color='orange', fill=True, alpha=0.5)
    ax.add_patch(circle)

# Dibuixem els VAPS:
plt.scatter(np.real(vaps),np.imag(vaps),color="black", s=5)

# Límits dels eixos:
plt.xlim(-10, 10)
plt.ylim(-10, 10)

# Labels dels eixos:
plt.xlabel('Part real')
plt.ylabel('Part imaginària')

plt.show()
```

```python
## Fem la intersecció entre la unió de discs per files i la unió de discs per columnes:

## Dibuixem:
fig, ax = plt.subplots()

ax.set_aspect('equal')  # Fem l'escala dels eixos iguals
ax.grid(True) # Afegim el grid


# Anem dibuixant cada disc (per files):
for i in range(len(A)):
    circle = plt.Circle((np.real(c[i]), np.imag(c[i])), r[i], color='blue', fill=True, alpha=0.3)
    ax.add_patch(circle)

# Anem dibuixant cada disc (per columnes):
for i in range(len(A)):
    circle = plt.Circle((np.real(c2[i]), np.imag(c2[i])), r2[i], color='orange', fill=True, alpha=0.3)
    ax.add_patch(circle)

# Dibuixem els VAPS:
plt.scatter(np.real(vaps),np.imag(vaps),color="black", s=5)

# Límits dels eixos:
plt.xlim(-10, 10)
plt.ylim(-10, 10)

# Labels dels eixos:
plt.xlabel('Part real')
plt.ylabel('Part imaginària')

plt.show()
```
