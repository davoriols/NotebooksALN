# Resolució de sistemes lineas amb mètodes iteratius

## Conceptes inicials:

Sigui $A\in \mathcal{M}_n(\mathbb{R})$ una matriu de rang màxim i sigui $\boldsymbol{b}$ un vector de $\mathbb{R}^n$. Un **mètode iteratiu** per resoldre el sistema lineal $A\boldsymbol{x} = \boldsymbol{b}$ és un procediment que permet construir una successió de vectors $\{\boldsymbol{x}^{(k)}\}_{k\geq0}$ tal que

$$
    \lim_{k\to+\infty} \boldsymbol{x}^{(k)} = A^{-1}\boldsymbol{b}.
$$

&nbsp;

En particular, direm que un mètode iteratiu és:

* **consistent** quan el límit de la sucesió anterior, en cas d'existir, és solució del sistema lineal.

* **convergent** si per qualsevol vector inicial $\boldsymbol{x}^{(0)}\in\mathbb{R}^n$ tenim que
$$
    \lim_{k\to+\infty} \boldsymbol{x}^{(k)} = A^{-1}\boldsymbol{b}.
$$



### Mètodes iteratius clàssics:

Els mètodes iteratius clàssics són de la forma

$$
\left\{
\begin{aligned}
&\boldsymbol{x}^{(0)} \in \mathbb{R}^n,\\
&\boldsymbol{x}^{(k+1)} = B\boldsymbol{x}^{(k)} + \boldsymbol{c},\quad k\geq0
\end{aligned}
\right.
$$

on:

* $B\in\mathcal{M}_n(\mathbb{R})$ és una matriu,

* $\boldsymbol{c}\in\mathbb{R}^n$ és un vector,

els quals calcularem a partir del sistema que volguem resoldre.

&nbsp;

>**Proposició:**
>
>Un mètode iteratiu
$$\boldsymbol{x}^{(k+1)} = B\boldsymbol{x}^{(k)} + \boldsymbol{c}, \quad{ k \geq 0}$$
>és consistent per resoldre el sistema lineal $A\boldsymbol{x} = \boldsymbol{b}$ si i només si el sistema
$$A\boldsymbol{x} = \boldsymbol{b},$$
>és equivalent a
$$\boldsymbol{x} = B\boldsymbol{x} + \boldsymbol{c}.$$


&nbsp;

> ⚠️ **Observació:** això ens està dient que:
>
> * $(I-B)$ és una matriu invertible.
>
> * $\boldsymbol{c} = (I-B)\boldsymbol{x} = (I-B)A^{-1}\boldsymbol{b}$.




&nbsp;

### Convergència dels mètodes iteratius clàssics

Ens agradaria poder evitar aplicar un mètode iteratiu per resoldre un sistema lineal si aquest no és convergent.

&nbsp;

Per poder fer-ho podem utilitzar els següents resultats:





##### **Teorema**

Un mètode iteratiu $\boldsymbol{x}^{(k+1)} = B\boldsymbol{x}^{(k)} + \boldsymbol{c}$ consistent pel sistema lineal $A\boldsymbol{x} = \boldsymbol{b}$  és convergent si i només si

$$
    \lim_{k\to+\infty} B^k \boldsymbol{y} = \boldsymbol{0}, \qquad \forall \boldsymbol{y}\in\mathbb{R}^n.
$$


&nbsp;



###### *Demostració*:

Donat que el mètode és consistent, sabem que:

$$ A\boldsymbol{x} = \boldsymbol{b} \Longleftrightarrow \boldsymbol{x} = B\boldsymbol{x} + \boldsymbol{c}$$

D'aquesta forma, emprant

* $\boldsymbol{x}^{(k+1)} = B\boldsymbol{x}^{(k)} + \boldsymbol{c}$,
* $\boldsymbol{x} = B\boldsymbol{x} + \boldsymbol{c}$,


tenim:

$$ \boldsymbol{x}^{(k+1)} - \boldsymbol{x} = B(\boldsymbol{x}^{(k)} - \boldsymbol{x}), \qquad \forall k \geq 0.$$

Emprant aquesta recurrència, tenim:

$$\boldsymbol{x}^{(k)} - \boldsymbol{x} = B(\boldsymbol{x}^{(k-1)} - \boldsymbol{x}) = B^2(\boldsymbol{x}^{(k-2)} - \boldsymbol{x}) = \ldots = B^k(\boldsymbol{x}^{(0)} - \boldsymbol{x}), \qquad \forall k\geq 0.$$

Recordem que per definició tenim que el mètode és convergent si i només si:

$$
\lim_{k\to+\infty} \boldsymbol{x}^{(k)} = \boldsymbol{x}, \quad \forall \boldsymbol{x}^{(0)} \in \mathbb{R}^n,
$$

que és equivalent a

$$
\lim_{k\to+\infty} \boldsymbol{x}^{(k)} - \boldsymbol{x}= \boldsymbol{0}, \quad \forall \boldsymbol{x}^{(0)} \in \mathbb{R}^n,
$$

i, per tant, si i només si

$$
\lim_{k\to+\infty} B^k(\boldsymbol{x}^{(0)} - \boldsymbol{x}) = \boldsymbol{0}, \quad \forall \boldsymbol{x}^{(0)} \in \mathbb{R}^n,
$$

el que equival a

$$
\lim_{k\to+\infty} B^k\boldsymbol{y} = \boldsymbol{0}, \quad \forall \boldsymbol{y} \in \mathbb{R}^n.
$$

##### **Teorema**

Sigui $B\in\mathcal{M}_n(\mathbb{R})$, aleshores les següents afirmacions són equivalents:



1.   $\displaystyle{\lim_{k\to+\infty} B^k\boldsymbol{y} = \boldsymbol{0},\quad \forall \boldsymbol{y}\in\mathbb{R}^n}$.

2.   $\displaystyle{\lim_{k\to+\infty} B^k = (0)}$, (matriu nul·la).

3. $\rho(B) < 1 $.

4. Per a alguna norma matricial subordinada es té $\|B\| < 1$.



&nbsp;

> ⚠️ **Atenció!**
>
> Els punts 3 i 4 ens poden ser de gran utilitat per assegurar la convergència d'un mètode iteratiu.

$$

### Mètodes de descomposició

Els mètodes iteratius clàssics que veurem són un exemple de mètodes de descomposició, ja que tots ells s'obtenen a partir d'una descomposició de la matriu del sistema lineal $A\boldsymbol{x}=\boldsymbol{b}$ de la forma:

$$A = M - N, $$

on $M$ acostuma a ser una matriu que podem invertir fàcilment. Així el sistema lineal anterior, és equivalent a:

$$
\begin{aligned}
    A\boldsymbol{x} = \boldsymbol{b} &\Longleftrightarrow (M-N)\boldsymbol{x} = \boldsymbol{b}\\
    &\Longleftrightarrow M\boldsymbol{x}-N\boldsymbol{x} = \boldsymbol{b}\\
    &\Longleftrightarrow M\boldsymbol{x} = N\boldsymbol{x} + \boldsymbol{b}\\
    &\Longleftrightarrow \boldsymbol{x} = \underbrace{M^{-1}N}_{B}\boldsymbol{x} + \underbrace{M^{-1}\boldsymbol{b}}_{\boldsymbol{c}}\\
\end{aligned}    
$$

i, per tant, el mètode iteratiu es pot expressar de la següent forma:

$$
\left\{
\begin{aligned}
&\boldsymbol{x}^{(0)} \in \mathbb{R}^n,\\
&\boldsymbol{x}^{(k+1)} = M^{-1}N\boldsymbol{x}^{(k)} + M^{-1}\boldsymbol{b},\quad k\geq0
\end{aligned}
\right.
$$

A la pràctica, moltes vegades ens serà convenient simplement treballar amb l'expressió:
$$
M\boldsymbol{x}^{(k+1)} = N\boldsymbol{x}^{(k)} +\boldsymbol{b}
$$

## Mètode iteratiu de Jacobi

El mètode iteratiu de Jacobi és un algorisme utilitzat per resoldre sistemes d'equacions lineals, en especial aquells que són estrictament dominants en la seva diagonal. Aquest mètode es basa en la descomposició de la matriu del sistema en components diagonal, triangular inferior i triangular superior. Per a cada iteració, es calcula una nova aproximació de la solució fins que es converteixi.

Aquest mètode **únicament es pot aplicar a sistemes on tots els seus elements diagonals siguin diferents de 0**.

Així, donat el sistema lineal $A\boldsymbol{x} = \boldsymbol{b}$, definim les matrius:

* $\displaystyle{L=\begin{pmatrix}0 &  &  &  & \\a_{2,1} &0 & & & \\ \vdots & \ddots & \ddots & \\ \vdots & & \ddots & 0 & \\ a_{n,1} & \cdots & \cdots & a_{n,n-1}& 0\end{pmatrix}}$

* $\displaystyle{D=\begin{pmatrix}a_{1,1} \\ & a_{2,2}\\ & & \ddots\\ & & & a_{n,n}\end{pmatrix}}$

* $\displaystyle{U=\begin{pmatrix}0 & a_{1,2} & \cdots & \cdots & a_{1,n}\\ &0 &\ddots & & \vdots \\& & \ddots & \ddots & \vdots\\ & & & 0 & a_{n-1,n}\\ & & & & 0\end{pmatrix}}$


&nbsp;

El mètode de Jacobi ve donat per:

$$
    \boldsymbol{x}^{(k+1)} = B_{J}\boldsymbol{x}^{(k)} + \boldsymbol{c}_J,\quad k\geq 0
$$

amb

* $B_J = -D^{-1}(L+U)$.

* $\boldsymbol{c}_J = D^{-1}\boldsymbol{b}$.


&nbsp;

Fixem-nos que el mètode de Jacobi és un mètode de descomposició amb:

* $M = D$

* $N = -(L+U)$

##### Construcció del mètode:

Fixem-nos que com $A = L+D+U$ i $A\boldsymbol{x} = \boldsymbol{b}$ tenim:

$$
\begin{aligned}
A\boldsymbol{x} = \boldsymbol{b} & \Longleftrightarrow (L+D+U)\boldsymbol{x} = \boldsymbol{b}\\
& \Longleftrightarrow D\boldsymbol{x}+(L+U)\boldsymbol{x} = \boldsymbol{b}\\
& \Longleftrightarrow D\boldsymbol{x} = -(L+U)\boldsymbol{x} + \boldsymbol{b}\\
&\Longleftrightarrow \boldsymbol{x} = \underbrace{-D^{-1}(L+U)}_{B_J}\boldsymbol{x} + \underbrace{D^{-1}\boldsymbol{b}}_{\boldsymbol{c}_J}
\end{aligned}
$$

i, per tant, tenim:

$$
\boldsymbol{x}^{(k+1)} = \underbrace{-D^{-1}(L+U)}_{B_J}\boldsymbol{x}^{(k)} + \underbrace{D^{-1}\boldsymbol{b}}_{\boldsymbol{c}_J}.
$$

&nbsp;

> **Observació:**
>
> * $\displaystyle{D^{-1}=\begin{pmatrix}\cfrac{1}{a_{1,1}} \\ & \cfrac{1}{a_{2,2}}\\ & & \ddots\\ & & & \cfrac{1}{a_{n,n}}\end{pmatrix}}$


&nbsp;

Component a component el que tenim és:


$$
    x_i^{(k+1)} = \frac{1}{a_{ii}}\left(b_i - \sum_{j\neq i}a_{ij}x_j^{(k)}\right),\qquad \text{ per } i = 1\div n.
$$

##### Convergència del mètode:

El mètode de Jacobi és convergent si la matriu $A$ és estrictament diagonalment dominant, i pot convergir inclòs si aquesta condició no es satisfà.

&nbsp;

### *Exercici:*

*Implementa una funció que, donada la matriu $A$ i el vector $\boldsymbol{b}$ del sistema $A\boldsymbol{x} = \boldsymbol{b}$ i **un vector inicial** $\boldsymbol{x}^{(0)}$ calculi la solució d'aquest sistema emprant el mètode de Jacobi.*

*Utilitzeu els tres paràmetres opcionals següents:*

* *`abstol=1.e-10`, que ens indiqui la tolerància mínima amb la qual volem la solució utilitzant la norma euclídea: $\|\boldsymbol{x}^{(k+1)}-\boldsymbol{x}^{(k)}\| < \varepsilon$. Un cop assolida, el mètode ha de retornar la solució.*

* *`reltol=1.e-10`, que ens indiqui la tolerància mínima amb la qual volem la solució utilitzant la norma euclídea: $\|\boldsymbol{x}^{(k+1)}-\boldsymbol{x}^{(k)}\|/\|\boldsymbol{x}^{(k+1)}\| < \varepsilon$. Un cop assolida, el mètode ha de retornar la solució.*

* *`maxIter=100`, que controli el nombre màxim d'iteracions possibles.*

*La funció ha de retornar la solució `x` obtinguda i el nombre d'iteracions realitzades `iter`, en aquest ordre. Si no s'ha assolit la tolerància absoluta `iter` serà -1, si no s'ha assolit la tolerància relativa -2, i si no s'han asolit cap de les dues -3.*

```python
### Espai per fer l'exercici
import numpy as np

def jacobi(A, b, x, abstol=1.e-10, reltol=1.e-10, maxIter=100):

    L = np.tril(A, -1)
    D = np.diag(1/np.diag(A))
    U = np.triu(A, 1)

    xold = x.copy()

    for iter in range(maxIter):

        x = -D@(L + U)@xold + D@b

        absolut = np.linalg.norm(x - xold)
        relatiu = absolut / np.linalg.norm(x)

        if (absolut < abstol and relatiu < reltol):
            return x, iter + 1

        xold = x.copy()

    if (absolut >= abstol and relatiu >= reltol):
        return x, -3

    elif (relatiu > reltol):
        return x, -2

    else:
        return x, -1

A = np.array([[2,1],
              [5,7]], dtype=np.float64)

b = np.array([11, 13], dtype=np.float64)
xold = np.array([1, 1], dtype=np.float64)


x, iter = jacobi(A, b, xold)

print(x)
print(iter)



```

### Mètode iteratiu de Gauss-Seidel

El mètode de Guass-Seidel és un altre mètode iteratiu clàssic de descomposició que té:

* $M = D + L$

* $N = -U$

on $L$, $D$ i $U$ són les matrius anteriors.

&nbsp;

Així, el mètode es pot escriure com:

$$
\begin{aligned}
A\boldsymbol{x} = \boldsymbol{b} & \Longleftrightarrow (L+D+U)\boldsymbol{x} = \boldsymbol{b}\\
& \Longleftrightarrow (L+D)\boldsymbol{x}+U\boldsymbol{x} = \boldsymbol{b}\\
& \Longleftrightarrow (L+D)\boldsymbol{x} = -U\boldsymbol{x} + \boldsymbol{b}\\
&\Longleftrightarrow \boldsymbol{x} = \underbrace{-(L+D)^{-1}U}_{B_{GS}}\boldsymbol{x} + \underbrace{(L+D)^{-1}\boldsymbol{b}}_{\boldsymbol{c}_{GS}}
\end{aligned}
$$

i, per tant, tenim:

$$
\boldsymbol{x}^{(k+1)} = \underbrace{-(L+D)^{-1}U}_{B_{GS}}\boldsymbol{x}^{(k)} + \underbrace{(L+D)^{-1}\boldsymbol{b}}_{\boldsymbol{c}_{GS}}.
$$


&nbsp;

Notem que, de la mateixa manera que pel cas del mètode de Jacobi, aquest mètode **únicament es pot aplicar a sistemes on tots els seus elements diagonals siguin diferent de 0**.

&nbsp;

Notem també que, a efectes d'implementació, no cal calcular la matriu inversa $(L+D)^{-1}$, ja que simplement el que estem fent és utilitzar els elements $x_1^{(k+1)},\ldots,x_{i-1}^{(k+1)}$ calculats en el càlcul de $x_{i}^{(k+1)}$. És a dir, component a component, tenim:

* $\displaystyle{x_1^{(k+1)} = \frac{1}{a_{11}}\left(b_1 -\sum_{j=2}^n a_{1j}x_j^{(k)}\right) }$

* $\displaystyle{x_i^{(k+1)} = \frac{1}{a_{ii}} \left(b_i - \sum_{j=1}^{i-1} a_{ij}x_j^{(k+1)} - \sum_{j=i+1}^{n} a_{ij}x_j^{(k)}\right)}$ per $i=2\div n-1$

* $\displaystyle{x_n^{(k+1)} = \frac{1}{a_{nn}}\left(b_n -\sum_{j=1}^{n-1} a_{nj}x_j^{(k+1)}\right) }$



#### Convergència del mètode:

El mètode de Gauss-Seidel és convergent si la matriu $A$ és estrictament diagonalment dominant (veure exercici 6), i pot convergir inclòs si aquesta condició no es satisfà.

### *Exercici*

*Implementa una funció que, donada la matriu $A$ i el vector $\boldsymbol{b}$ del sistema $A\boldsymbol{x} = \boldsymbol{b}$ i **un vector inicial** $\boldsymbol{x}^{(0)}$ calculi la solució d'aquest sistema emprant el mètode de Gauss-Seidel. **Eviteu calcular la matriu inversa de** $(L+D)$.*

*Utilitzeu els tres paràmetres opcionals següents:*

* *`abstol=1.e-10`, que ens indiqui la tolerància mínima amb la qual volem la solució utilitzant la norma euclídea: $\|\boldsymbol{x}^{(k+1)}-\boldsymbol{x}^{(k)}\| < \varepsilon$. Un cop assolida, el mètode ha de retornar la solució.*

* *`reltol=1.e-10`, que ens indiqui la tolerància mínima amb la qual volem la solució utilitzant la norma euclídea: $\|\boldsymbol{x}^{(k+1)}-\boldsymbol{x}^{(k)}\|/\|\boldsymbol{x}^{(k+1)}\| < \varepsilon$. Un cop assolida, el mètode ha de retornar la solució.*

* *`maxIter=100`, que controli el nombre màxim d'iteracions possibles.*

*La funció ha de retornar la solució `x` obtinguda i el nombre d'iteracions realitzades `iter`, en aquest ordre. Si no s'ha assolit la tolerància absoluta `iter` serà -1, si no s'ha assolit la tolerància relativa -2, i si no s'han asolit cap de les dues -3.*

```python
### Espai per fer l'exercici
import numpy as np
from scipy.linalg import solve_triangular

def gauSei(A, b, x, abstol=1.e-10, reltol=1.e-10, maxIter=100):

    L = np.tril(A, -1)
    D = np.diag(np.diag(A))
    U = np.triu(A, 1)

    xold = x.copy()

    for iter in range(maxIter):

        Uxb = -U@xold + b

        x = solve_triangular(L + D, Uxb, lower = True)

        absolut = np.linalg.norm(x - xold)
        relatiu = absolut / np.linalg.norm(x)

        if (absolut < abstol and relatiu < reltol):
            return x, iter + 1

        xold = x.copy()

    if (absolut >= abstol and relatiu >= reltol):
        return x, -3

    elif (relatiu >= reltol):
        return x, -2

    else:
        return x, -1

A = np.array([[16,3],
              [7,-11]], dtype=np.float64)

b = np.array([11, 13], dtype=np.float64)
xold = np.array([1, 1], dtype=np.float64)


x, iter = gauSei(A, b, xold)

print(x)
print(iter)


```

### *Exercici 4:*

*Donat el sistema lineal $A\boldsymbol{x}=\boldsymbol{b}$ amb*

$$
A=\begin{pmatrix}
3 & 2\\ 2 & 6
\end{pmatrix}
$$

*justifiqueu si el mètode de Gauss-Seidel és convergent.*

#### Resolució:

Com la matriu $A$ és estrictament diagonalment dominant sabem que el mètode de Gauss-Seidel serà convergent.

Veiem aquesta proposició:

&nbsp;

##### **Proposició:**

Si un sistema lineal $A\boldsymbol{x} = \boldsymbol{b}$ és estrictament diagonalment dominant, aleshores el mètode de Gauss-Seidel és convergent.


###### Demostració:

Recordem que la matriu del mètode de Gauss-Seidel ve donada per:

* $B_{GS} = -(L+D)^{-1}U.$

I, sabem que, un mètode serà convergent si el radi espectral de la matriu del mètode és menor que $1$. Així doncs, el que voldrem veure és que si la matriu $A$ és estrictament diagonalment dominant aleshores $\rho(B_{GS}) < 1$.

Sigui $\lambda$ un VAP qualsevol de $B_{GS}$ amb VEP associat $\boldsymbol{v}$. Aleshores

$$ B_{GS} \boldsymbol{v} = -(L+D)^{-1}U \boldsymbol{v} = \lambda \boldsymbol{v} \Longleftrightarrow - U\boldsymbol{v} = \lambda(L+D) \boldsymbol{v}$$

És a dir, tenim:

$$
- \sum_{j=i+1}^n a_{ij}v_j = \lambda \sum_{j=1}^i a_{ij}v_j =  \lambda \sum_{j=1}^{i-1} a_{ij}v_j + \lambda a_{ii} v_i, \qquad \forall i = 1,\ldots, n.
$$

D'aquesta forma, tenim:

$$
\lambda a_{ii} v_i = - \sum_{j=i+1}^n a_{ij}v_j -  \lambda \sum_{j=1}^{i-1} a_{ij}v_j, \qquad \forall i = 1,\ldots, n.  
$$

Prenent valor absolut a banda i banda, tenim:

$$
\begin{aligned}
|\lambda|\,|a_{ii}|\,|v_i| &= |\lambda a_{ii} v_i|\\
& = \left|- \sum_{j=i+1}^n a_{ij}v_j -  \lambda \sum_{j=1}^{i-1} a_{ij}v_j\right|\\
& \leq \sum_{j=i+1}^n |a_{ij}|\,|v_j| +  |\lambda| \sum_{j=1}^{i-1} |a_{ij}|\,|v_j|,\qquad \forall i = 1,\ldots,n
\end{aligned}
$$

Imposem que $\|\boldsymbol{v}\|_\infty = 1$ (triem el VEP de manera que la màxima coordenada sigui $\pm1$) i, suposem que la coordenada més gran en valor absolut és la $k$-èssima, és a dir, tenim:

* $|v_k| = 1$.

* $|v_j| \leq 1,\, \forall j=1,\ldots, n$.

Així, tenim:

$$
\begin{array}{lll}
|\lambda|\,|a_{kk}| &= |\lambda|\, |a_{kk}|\,|v_k| & \quad \big(|v_k| = 1\big)\\
& \leq \displaystyle{\sum_{j=k+1}^n |a_{kj}|\,|v_j| +  |\lambda| \sum_{j=1}^{k-1} |a_{kj}|\,|v_j|} & \quad \big(\text{triem }k=i\big)\\
& \leq \displaystyle{\sum_{j=k+1}^n |a_{kj}| +  |\lambda| \sum_{j=1}^{k-1} |a_{kj}|} & \quad \big(|v_j| \leq 1\big)\\
\end{array}
$$

Així:

$$
|\lambda|\left(|a_{kk}| - \sum_{j=1}^{k-1} |a_{kj}| \right) \leq \sum_{j=k+1}^n |a_{kj}|.
$$

i d'aquesta manera:

$$
    |\lambda| \leq \cfrac{\displaystyle{\sum_{j=k+1}^n |a_{kj}|}}{\displaystyle{|a_{kk}| - \sum_{j=1}^{k-1} |a_{kj}|}} < 1.
$$

on hem emprat que $A$ és estrictament diagonalment dominant i, per tant, tenim:

$$
    |a_{kk}| > \sum_{j=1}^{k-1} |a_{kj}| + \sum_{j=k+1}^n |a_{kj}| \Longleftrightarrow |a_{kk}| - \sum_{j=1}^{k-1} |a_{kj}| > \sum_{j=k+1}^n |a_{kj}| \geq 0.
$$






> ##### **Observació 1:**
>
> La demostració també ens permet veure que és cert per columnes (considerariem els VEPs de la trasposada).

&nbsp;

> ##### **Observació 2:**
>
> La demostració pel cas de Jacobi és pot fer de manera anàloga, però és molt més senzill utilitzar directament que la norma $\|B_J\|_1< 1$ o la norma $\|B_J\|_\infty <1$.






### *Exercici 6:*

*Considereu el sistema lineal amb matriu*

$$
A=
\begin{pmatrix}
\alpha &  0     &  1   \\
  0    & \alpha &  0   \\
  1    &  0     & \alpha
\end{pmatrix},
\qquad \alpha\in \mathbb{R}.
$$

*Analitzeu la convergència dels mètodes de Jacobi i de Gauss-Seidel per a la seva resolució.*

#### Resolució:


Sabem que si $|\alpha| > 1$ ambdos mètodes seran convergents, però, és optimal aquest resultat? Veiem-ho:

&nbsp;

**Jacobi:**

Recordem que la matriu del mètode de Jacobi ($B_J$) té per elements:

$$
(B_{ij}) = \left\{\begin{array}{rl}-\cfrac{a_{ij}}{a_{ii}}& \quad\text{si }i\neq j\\ 0 &\quad\text{si }i= j& \end{array}\right.
$$
així, necessitem que $\alpha \neq 0$ i tenim:

$$
    B_J = \begin{pmatrix} 0 & 0 & -\cfrac{1}{\alpha}\\ 0 & 0 & 0\\ -\cfrac{1}{\alpha} & 0 & 0 \end{pmatrix}.
$$

Els valors propis de $B_J$ són les arrels de $\det(B_J-\lambda Id) = 0$:

$$
  0  = \det(B_J-\lambda Id) = \begin{vmatrix} -\lambda & 0 & -\cfrac{1}{\alpha}\\ 0 & -\lambda & 0\\ -\cfrac{1}{\alpha} & 0 & -\lambda \end{vmatrix} = -\lambda\left(\lambda^2-\cfrac{1}{\alpha^2}\right) = -\lambda\left(\lambda-\cfrac{1}{\alpha}\right)\left(\lambda+\cfrac{1}{\alpha}\right)
$$

Per tant, els valor propis són $0$, $\cfrac{1}{\alpha}$ i $-\cfrac{1}{\alpha}$. Com necesistem que el més gran en valor absolut sigui menor que $1$ tenim $|\alpha| > 1$.

&nbsp;

**Gauss-Seidel:**

Recordem que la matriu del mètode Gauss-Seidel ve donada per:

$$\begin{aligned}B_{GS} &= -(L+D)^{-1}U = -\begin{pmatrix}\alpha & 0 & 0\\ 0 & \alpha & 0\\ 1 & 0&\alpha\end{pmatrix}^{-1}\begin{pmatrix}0 & 0 & 1\\ 0 & 0 & 0\\ 0 & 0&0\end{pmatrix}\\ &= -\begin{pmatrix}\cfrac{1}{\alpha} & 0 & 0\\ 0 & \cfrac{1}{\alpha} & 0\\ -\cfrac{1}{\alpha^2} & 0&\cfrac{1}{\alpha}\end{pmatrix}\begin{pmatrix}0 & 0 & 1\\ 0 & 0 & 0\\ 0 & 0&0\end{pmatrix} = \begin{pmatrix}0 & 0 & -\cfrac{1}{\alpha}\\ 0 & 0 & 0\\ 0 & 0&\cfrac{1}{\alpha^2}\end{pmatrix}\end{aligned}$$

&nbsp;

Els valors propis de $B_{GS}$ són les arrels de $\det(B_{GS}-\lambda Id) = 0$:

$$
  0  = \det(B_{GS}-\lambda Id) = \begin{vmatrix} -\lambda & 0 & -\cfrac{1}{\alpha}\\ 0 & -\lambda & 0\\ 0 & 0 & \cfrac{1}{\alpha^2}-\lambda \end{vmatrix} = -\lambda^2\left(\lambda-\cfrac{1}{\alpha^2}\right)
$$


Per tant, els valor propis són $0$ i $\cfrac{1}{\alpha^2}$. Com necesistem que el més gran en valor absolut sigui menor que $1$ tenim $|\alpha| > 1$.

### *Exercici 7:*

*Doneu condicions suficients sobre el paràmetre $\beta$ per tal que els mètodes de Jacobi i de Gauss-Seidel convergeixin en aplicar-los per trobar la solució del sistema que té per matriu*

$$
A=
\begin{pmatrix}
-10    &  2    \\
\beta  &  5  
\end{pmatrix}.
$$

#### Resolució:

Pel que hem vist abans:



*   $A$ estrictament diagonal dominant per files quan $|\beta|< 5$.
*   $A$ estrictament diagonal dominant per columnes quan $|\beta|< 10$.

Així, és suficient que $|\beta|< 10$ per garantir la convergència. Ara bé, és un valor òptim?

&nbsp;

**Jacobi:**

* $\displaystyle{B_J = \begin{pmatrix}0 & \cfrac{1}{5}\\ -\cfrac{\beta}{5} & 0\end{pmatrix}}$.

Ara, calculem els valors propis i veiem per quins valors de $\beta$ tenim $\rho(B_J) < 1$:

* $\displaystyle{0 = \det(B_J - \lambda Id) = \begin{vmatrix}-\lambda & \cfrac{1}{5}\\ -\cfrac{\beta}{5} & -\lambda\end{vmatrix} = \lambda^2+\frac{\beta}{25} = \left(\lambda-\frac{\sqrt{\beta}}{5}\right)\left(\lambda+\frac{\sqrt{\beta}}{5}\right)}$

Així doncs, els valors propis són $\pm\cfrac{\sqrt{\beta}}{5}$ i tenim $\rho(B_J) = \cfrac{\sqrt{|\beta|}}{5}$. Per tant, necessitem $|\beta| < 25$.

&nbsp;

**Gauss-Seidel:**

Pel mètode de Gauss-Seidel, tenim:

$$
B_{GS} = -(L+D)^{-1}U = - \begin{pmatrix} -10    &  0    \\ \beta  &  5  \end{pmatrix}^{-1}\begin{pmatrix}0 & 2\\ 0 & 0\end{pmatrix} = -\frac{1}{50}\begin{pmatrix}-1 & 0 \\ \beta & 10\end{pmatrix}\begin{pmatrix}0 & 2\\ 0 & 0\end{pmatrix} = \begin{pmatrix} 0 & \cfrac{1}{5}\\ 0 & -\cfrac{\beta}{25}\end{pmatrix}
$$


Ara, calculem els valors propis i veiem per quins valors de $\beta$ tenim $\rho(B_{GS}) < 1$:

* $\displaystyle{0 = \det(B_{GS} - \lambda Id) = \begin{pmatrix} -\lambda & \cfrac{1}{5}\\ 0 & -\cfrac{\beta}{25}-\lambda\end{pmatrix} = \lambda\left(\lambda + \cfrac{\beta}{25}\right)}$

Així doncs, els valors propis són $0$ i $\cfrac{\beta}{25}$ i tenim $\rho(B_J) = \cfrac{|\beta|}{25}$. Per tant, necessitem $|\beta| < 25$.
