### FÃģrmules de propagaciÃģ dels errors:

Per problemes numÃĻrics on calculem $y=f(x)$ a partir d'una Ãšnica dada $x$ tenim les segÃžents fÃģrmules aproximades de propagaciÃģ de l'error:

$$ e_a(y) \approx f'(x)\cdot e_a(x),$$

$$ e_r(y) \approx \frac{f'(x)\cdot x}{f(x)}\,e_r(x),$$

on $e_a$ indica l'error absolut i $e_r$ l'error relatiu.  D'igual forma, tenim les fÃģrmules aproximades de propagaciÃģ de l'error maximal:

$$ \varepsilon_a(y) \approx |f'(x)|\cdot \varepsilon_a(x),$$

$$ \varepsilon_r(y) \approx \frac{|f'(x)\cdot x|}{|f(x)|}\,\varepsilon_r(x).$$


&nbsp;

Per problemes numÃĻrics mÃĐs generals, en els quals hem de calcular un resultat $y=g(x_1,x_2,...,x_n)$ a partir de les dades $x_1$, $x_2$, $\ldots$, $x_n$ tenim la generalitzaciÃģ de les fÃģrmules aproximades de propagaciÃģ de l'error:

$$ e_a(y) \approx \sum_{i=1}^n \frac{\partial g}{\partial x_i}\,(x_1,x_2,\ldots,x_n)\cdot e_a(x_i),$$

$$ e_r(y) \approx \frac{1}{g(x_1,x_2,\ldots,x_n)}\sum_{i=1}^n \frac{\partial g}{\partial x_i}\,(x_1,x_2,\ldots,x_n)\cdot x_i \cdot e_r(x_i).$$

I, d'igual manera, a partir de les fites $\varepsilon_a(x_i)$ dels errors absoluts $e_a(x_i)$ tenim la generalitzaciÃģ de les fÃģrmules aproximades de propagaciÃģ de l'error maximal:

$$ \varepsilon_a(y) \approx \sum_{i=1}^n \left|\frac{\partial g}{\partial x_i}\,(x_1,x_2,\ldots,x_n)\right|\cdot \varepsilon_a(x_i),$$


$$ \varepsilon_r(y) \approx \frac{1}{|g(x_1,x_2,\ldots,x_n)|}\sum_{i=1}^n \left|\frac{\partial g}{\partial x_i}\,(x_1,x_2,\ldots,x_n)\right|\cdot |x_i| \cdot \varepsilon_r(x_i).$$



### Xifres significatives i decimals correctes

Sigui $\bar{x}$ una aproximaciÃģ de $x$.



*   Si $|e_a(\bar{x})| \leq \cfrac{1}{2}\, \cdot 10^{-k}$, direm que $\bar{x}$ tÃĐ $k$ **decimals correctes**.

*   Si $|e_r(\bar{x})| \leq \cfrac{1}{2}\, \cdot 10^{1-k}$, direm que $\bar{x}$ tÃĐ $k$ **xifres significatives**.



Aquest concepte tambÃĐ es pot extende pel cas de nombres en base $b$.

#### ResoluciÃģ

Comprovem els resultats obtinguts amb la llibreria simbÃēlica sympy:

```python
import sympy as sp

# Definim el problema original amb el valor simbÃēlic de sqrt(3)
real = (7-4*sp.sqrt(3))**4

# Calculem simbÃēlicament els resultats utilitzant l'aproximaciÃģ de 1.73205 per sqrt(3).
res1 = sp.simplify( 1/(7+4*1.73205)**4)
res2 = sp.simplify( 1/(18817+10864*1.73205))


# Printem el valor real (amb 15 xifres):
print("El valor real del problema amb 15 xifres ÃĐs ", real.evalf(15),"\n")

# Printem les aproximacions (amb 15 xifres):
print("L'aproximaciÃģ calculada amb la primera opciÃģ (amb 15 xifes) ÃĐs ", res1.evalf(15))
print("L'aproximaciÃģ calculada amb la segona opciÃģ (amb 15 xifes) ÃĐs ", res2.evalf(15),"\n")


# Calculem els errors:
e1_a = sp.Abs(real-res1)
e2_a = sp.Abs(real-res2)
e1_r = e1_a/sp.Abs(real)
e2_r = e2_a/sp.Abs(real)


# Printem els errors (amb 15 xifres):
print("El valor absolut de l'error absolut amb el primer mÃĻtode ÃĐs aproximadament", e1_a.evalf(15))
print("El valor absolut de l'error absolut amb el segon mÃĻtode ÃĐs aproximadament", e2_a.evalf(15), "\n")

print("El valor absolut de l'error relatiu amb el primer mÃĻtode ÃĐs aproximadament", e1_r.evalf(15))
print("El valor absolut de l'error relatiu amb el segon mÃĻtode ÃĐs aproximadament", e2_r.evalf(15), "\n")
```
