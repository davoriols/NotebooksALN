### Fórmules de propagació dels errors:

Per problemes numèrics on calculem $y=f(x)$ a partir d'una única dada $x$ obtenim la **fórmula aproximada de propagació de l'error**

$$ e_a(y) \approx f'(x)\cdot e_a(x),$$

on $e_a$ indica l'error absolut. Aquesta fórmula és conseqüència del [Teorema del valor mitjà](https://ca.wikipedia.org/wiki/Teorema_del_valor_mitj%C3%A0). D'igual forma, tenim la **fórmula aproximada de propagació de l'error maximal**

$$ \varepsilon_a(y) \approx |f'(x)|\cdot \varepsilon_a(x).$$


&nbsp;

Per problemes numèrics més generals, en els quals hem de calcular un resultat $y=g(x_1,x_2,...,x_n)$ a partir de les dades $x_1$, $x_2$, $\ldots$, $x_n$ tenim la generalització de la **fórmula aproximada de propagació de l'error**

$$ e_a(y) \approx \sum_{i=1}^n \frac{\partial g}{\partial x_i}\,(x_1,x_2,\ldots,x_n)\cdot e_a(x_i).$$

I, d'igual manera, a partir de les fites $\varepsilon_a(x_i)$ dels errors absoluts $e_a(x_i)$ tenim la generalització de la **fórmula aproximada de propagació de l'error maximal**

$$ \varepsilon_a(y) \approx \sum_{i=1}^n \left|\frac{\partial g}{\partial x_i}\,(x_1,x_2,\ldots,x_n)\right|\cdot \varepsilon_a(x_i).$$



## Càlcul simbòlic amb python

[SymPy](https://docs.sympy.org/latest/index.html) és una llibreria de Python per a matemàtiques simbòliques. Permet efectuar operacions matemàtiques amb símbols en lloc de valors numèrics, cosa que la fa útil per a càlculs exactes, manipulació d'expressions algebraiques, càlcul diferencial i integral, resolució d'equacions, etc.

Sympy, conté una multitud de funcions, algunes d'elles, a tall d'exemple, són:

| Comanda            | Descripció                                           |
|--------------------|------------------------------------------------------|
| `symbols('x y z')` | Crea les variables simbòliques `x`, `y` i `z`        |
| `sin(x)`           | Funció trigonomètrica sinus                          |
| `cos(x)`           | Funció trigonomètrica cosinus                        |
| `tan(x)`           | Funció trigonomètrica tangent                        |
| `sqrt(x)`          | Funció arrel quadrada                                |
| `log(x)`           | Funció logaritme natural                             |
| `simplify(expr)`   | Simplifica l’expressió `expr`                        |
| `expand(expr)`     | Expandeix l’expressió `expr`                         |
| `diff(f, x)`       | Calcula la derivada de `f` respecte a `x`            |
| `integrate(f, x)`  | Calcula la integral indefinida de `f` respecte a `x` |

Veiem un petit exemple de codi:

```python
from sympy import symbols, sin, cos, diff, integrate, pi, latex

# Creem la variable simbòlica x:
x = symbols('x')

# Funció:
f = sin(pi*x)*cos(pi*x)

# Calculem la derivada:
derivada = diff(f,x) # Podem calcular la derivada n-èssima mitjançant diff(f,x,n)

# Calculem la integral:
integral = integrate(f, x) # Podem calcular la integral entre a i b fent integrate(f,(x,a,b))

# Imprimim per pantalla el resultat:
print(f"La derivada de f(x) = {f} és {derivada}")
print(f"La integral de f(x) = {f} és {integral}")
```

```python
integral # No posem l'opció print perquè així surt en format LaTeX:
```

```python
display(integral) #Alteranativa per mostrar la fórmula en format LaTex
```

```python
print(latex(integral)) # I així podem imprimir el codi en LaTeX :)
```

### Exercici 9

*Demostreu que en l'operació $\sqrt{x}$ l'error relatiu és aproximadament la meitat de l'error relatiu en les dades. Direm que l'operació de fer $\sqrt{x}$ és una operació segura respecte de l'error relatiu. Feu patent la "inseguretat" de l'operació $f(x) = \cfrac{1}{1-x^{2}}$ per a $x\simeq 1$. (Suposeu que tenim errors només en la representació de les dades).*

&nbsp; &nbsp; &nbsp; *(a mà utilitzant la següent proposició)*

&nbsp;

#### Resolució

Abans de resoldre el problema veiem la següent proposició:

&nbsp;

**Proposició:**

Sigui $f$ de classe $C^1$ i considerem l'interval $(x,\bar{x})$ (o bé a l'interval $(\bar{x},x)$ depenent del valor de $\bar{x}$). Aleshores:

1.   $\varepsilon_a(f(\bar{x})) \approx \left| f'(\bar{x})\right|\cdot \varepsilon_a(\bar{x}),$

2.   $\varepsilon_r(f(\bar{x})) \approx \left| \cfrac{f'(\bar{x})\cdot\bar{x}}{f(\bar{x})}\right| \cdot \varepsilon_r(\bar{x}).$


Demostració:

$1)$ Utilitzarem el Teorema del Valor MItjà

$
\begin{array}{rll}
  |e_a(f(\bar{x}))| :=& |f(\bar{x})- f(x)|  &\Big(\text{Definició de l'error absolut}\Big)\\
   =& |f'(\xi_x)|\cdot|\bar{x}-x|, \text{ amb } \xi_x \in (x,\bar{x}) \text{ o } (\bar{x},x)\qquad &\Big(\text{Teorema del Valor Mitjà}\Big) \\
  = &|f'(\xi_x)|\cdot|e_a(\bar{x})|, \text{ amb } \xi_x \in (x,\bar{x}) \text{ o } (\bar{x},x) &\Big(\text{Definició de l'error absolut}\Big) \\
  \approx & |f'(\bar{x})|\cdot|e_a(\bar{x})| &\Big(\text{Continuïtat de }f'\Big) \\
\end{array}
$

I automàticament deduïm la relació per les fites ($\varepsilon_a$) de l'error absolut ($e_a$)

&nbsp;

$2)$ És conseqüència directa del resultat anterior:

$
\begin{array}{rll}
|e_r(f(\bar{x}))| := & \left| \cfrac{e_a(f(\bar{x}))}{f(x)}\right| & \Big(\text{Definició de l'error relatiu} \Big)\\
\approx &  \cfrac{|f'(\bar{x})|}{|f(x)|} \cdot |e_a(\bar{x})| &\Big( \text{Apartat }1\text{)}\Big)\\
\approx & \cfrac{|f'(\bar{x})|}{|f(\bar{x})|} \cdot |e_a(\bar{x})| &\Big( \text{Continuïtat de }f\Big)\\
= & \cfrac{|f'(\bar{x})|}{|f(\bar{x})|} \cdot|x| \cdot |e_r(\bar{x})| &\Big( |e_a(x)| = |x|\cdot |e_r(\bar{x})|\Big)\\
\approx  & \cfrac{|f'(\bar{x})|\cdot |\bar{x}|}{|f(\bar{x})|} \cdot |e_r(\bar{x})| \qquad &\Big( x \approx \bar{x}\Big)
\end{array}
$

I automàticament deduïm la relació per les fites ($\varepsilon_r$) de l'error relatiu ($e_r$).

$\square$

### Exercici 11

*Treballant amb $5$ xifres decimals, calculeu*
$$
	\sqrt[k]{2.15283} - \sqrt[k]{2.15263},\qquad\qquad k=2,3,4.
$$

$a$) *Directament.*

$b$) *Usant fórmules millors des del punt de vista numèric. (Indicació: Feu la divisió de polinomis $(a^{k} - b^{k})/(a-b)$).*

$c$) *Compareu els resultats i comenteu-los.*


#### Resolució: (Pissarra + codi)

$a$) Avaluem directament utilitzant la llibreria decimal:

```python
import decimal as dc

# Establir la precisió a 6 xifres significatives
dc.getcontext().prec = 6


# Nombres:
A = dc.Decimal('2.15283')
B = dc.Decimal('2.15263')

# Capçalera del resultat:
print(" k |  2.15283^(1/k)  | 2.15263^(1/k) | 2.15283^(1/k) - 2.15263^(1/k) ")
print("---------------------------------------------------------------------")

for i in range(2,5):
    k = dc.Decimal(i)

    # Calculem l'arrel k-èssima
    ak = A ** (dc.Decimal('1') / k)
    bk = B ** (dc.Decimal('1') / k)

    print(f" {k} |     {ak}     |    {bk}    |            {ak-bk}")
```

$b$) Pissarra + codi

```python
## Espai per fer l'exercici:

import decimal as dc

dc.getcontext().prec = 6 # Ho hem de tornar a posar en aquesta cel·la ja que el context només afecta a la cel·la

A = dc.Decimal('2.15283')
B = dc.Decimal('2.15263')
C = dc.Decimal('0.0002')


# Iterem en k:
for k in range(2,5):
    kDec = dc.Decimal(k)

    # Inicialitzem sumk, que contindrà el sumatori com B^{(k-1)/k}:
    sumk = B ** (dc.Decimal(k-1)/kDec)

    # Continuem sumant els termes A^{(i-1)/k}*B^{(i-1)/k} des de i = 2 fins a k-1:
    for i in range(2,k):
      sumk += A ** (dc.Decimal(i-1)/kDec) * B ** (dc.Decimal(k-i)/kDec)

    # Sumem per últim el terme A^{(k-1)/k}:
    sumk += A ** (dc.Decimal(k-1)/kDec)

    # Calculem el resultat final:
    res = C/sumk

    # Imprimim el resultat per pantalla:
    print(res)
```

### Exercici 14

*Usant un mètode recurrent, calculeu el valor de les integrals*
$$
	I_{j} = \int_{0}^{1} x^{j}\sin(\pi x)\,dx,
$$

*($j=2,4,\dots,20$). Estudieu l'estabilitat del mètode trobat.
__Nota:__ en general, direm que un mètode és __estable__ si esmorteeix els errors fets en les aproximacions. Quan, pel contrari, els amplifica, direm que el mètode és __inestable__*.

#### Resolució (Pissara + codi)

En aquest exercici ens serà de gran utilitat la llibreria sympy per calcular la integral de manera simbólica.

Si ho intentem fer directament... El programa triga molt!!

```python
import sympy as sp # Importem la llibreria sympy amb un àlies, moltes vegades s'importa sense cap mena d'àlies
import time # Amb aquesta llibreria podrem veure el temps d'execució del programa

# Temps inicial:
start_time = time.time()


######## Càlcul de la integral ########

# Definim els simbols, en aquest cas la variable x:
x = sp.symbols('x')

# Definim la funció f(x)
f = x**20 * sp.sin(sp.pi*x)

# Calculem la integral definida entre 0 i 1
integral = sp.integrate(f, (x, 0, 1))

########### Fi del programa ###########


# Temps final:
end_time = time.time()

# Calculem el temps total d'execució:
execution_time = end_time - start_time

# Printem la integral obtinguda:
print(f"La integral obtinguda és:\n {integral}\n")

# També podem printar la formula en latex:
display(integral)

# Printem el temps que hem trigat en calcular la integral simbòlica:
print(f"\nEl temps que hem necessitat per fer la integració simbòlica és {execution_time} segons.")

# Evaluem la integral amb una precisió de 15 dígits
resultat = integral.evalf(15)
print(f"El resultat aproximat de la integral és {resultat}")

```

**Recurrència 1:**

Obtinguda fent integració per parts.

$$
\left\{
\begin{aligned}
I_0 &= \frac{2}{\pi}\\
I_k &= \frac{1}{\pi} - \frac{k(k-1)}{\pi^2}I_{k-2},\qquad \text{ per } k =2,4,6,\ldots
\end{aligned}
\right.
$$

Aquesta manera de calcular la integral és molt lenta. Per optimitzar el procediment podem utilitzar la recurrència calculada.

```python
import sympy as sp

# Creem un vector simbòlic de 0's:
valors_integrals = [sp.symbols('0')]*11

# Incialitzem la primera posició amb el valor 2/pi
valors_integrals[0] = 2/sp.pi

u_div_pi = 1/sp.pi

for i in range(2,21,2):
    ind = i//2 # Index del vector
    valors_integrals[ind] = sp.simplify(u_div_pi - i*(i-1)/(sp.pi)**2*valors_integrals[ind-1])

valors_integrals[10]
```

Utilitzant numpy fes un programa que calculi la integral numèricament utilitzant precisió simple fent servir la recurrència anterior i els guardi en un vector anomenat `valors_rec_dir` on la primera posició coorespon a $I_0$, la segona a $I_2$, la tercera a $I_4$, etc.

```python
## Espai per fer l'exercici:

import numpy as np

# Vectors de valors inicials:
valors_rec_dir = np.zeros(11, dtype = np.float32)

```

Comparem els resultats obtinguts:

```python
# Printem els resultats obtinguts:
print("|  k |      I_k      | I_k recurrència 1 |    abs(e_r)    |")
print("|----|---------------|-------------------|----------------|")

for i in range(0,11):
    val_int_i = valors_integrals[i].evalf(15)
    print(f"| {2*i:2d} | {val_int_i:>13.7e} |  {valors_rec_dir[i]:>14.7e}   |",
           f"{np.abs((valors_rec_dir[i]-val_int_i)/val_int_i):>13.7e}  |")
```
