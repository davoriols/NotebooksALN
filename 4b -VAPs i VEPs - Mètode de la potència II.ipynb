# VAPs i VEPs: Mètode de la potència II

## Quocient de Rayleigh i mètode de la potència

Recordem breument els conceptes que ja havíem vist: el quocient de Rayleigh i el mètode de la potència.

##### **Definició:** quocient de Rayleigh

Donat un vector $\boldsymbol{x}\in \mathbb{C}^n\setminus\{ \boldsymbol{0}\}$ i $A\in\mathcal{M}_n(\mathbb{C})$, definim el seu quocient de Rayleigh com:

$$
    \mathcal{R}(A,\boldsymbol{x}) = \frac{\boldsymbol{x}^* A\boldsymbol{x}}{\boldsymbol{x}^* \boldsymbol{x}}
$$



Basats en aquesta idea, notem que, donada la matriu $A$, si $\boldsymbol{v}$ és un vector propi de valor propi $\lambda$, tenim:

$$
 A\boldsymbol{v} = \lambda \boldsymbol{v}
$$

Així:

$$
    \frac{\boldsymbol{v}^* A\boldsymbol{v}}{\boldsymbol{v}^* \boldsymbol{v}} = \frac{\boldsymbol{v}^* \lambda\boldsymbol{v}}{\boldsymbol{v}^* \boldsymbol{v}} = \lambda\frac{\boldsymbol{v}^* \boldsymbol{v}}{\boldsymbol{v}^* \boldsymbol{v}} = \lambda
$$

I, per tant, donat un vector propi, podem determinar el seu valor propi fàcilment.

&nbsp;

### Mètode de la potència

El mètode de la potència és, probablement, el mètode més senzill per calcular el VAP de mòdul màxim i el seu VEP associat.

Suposem $A\in\mathcal{M}_n(\mathbb{C})$ una matriu amb autovalors $\lambda_1, \lambda_2,\ldots,\lambda_n$ que suposarem ordenats de la forma

$$|\lambda_1| \geq |\lambda_2| \geq \ldots |\lambda_n|.$$

L'autovalor amb mòdul més gran es diu *dominant*, en cas de tenir $|\lambda_1| > |\lambda_2|$ direm que és *estrictament dominant*.

#### **Mètode de la potència iterada amb quocient de Rayleigh: (No l'utilitzem a la pràctica)**



El mètode es basa en la iteració següent:

$$
\left\{
\begin{aligned}&\boldsymbol{z}_0\in\mathbb{C}^n\setminus\{\boldsymbol{0}\},\\& \boldsymbol{z}_{k+1} = A\boldsymbol{z}_{k}, \,\,k\geq 0,\end{aligned}
\right.
$$

i en el càlcul dels termes successius $\{\sigma_k\}_{k\geq 0}$:

$$
\sigma_k = \frac{\boldsymbol{z}_k^*A\boldsymbol{z}_k}{\boldsymbol{z}_k^*\boldsymbol{z}_k} = \frac{\boldsymbol{z}_k^*\boldsymbol{z}_{k+1}}{\boldsymbol{z}_k^*\boldsymbol{z}_k},
$$

que tendiran a l'autovalor dominant $\lambda_1$.

&nbsp;

**Observació:** També es pot obtenir el valor de $\lambda_1$ dividint alguna de les component no nules de $\boldsymbol{z}_{k+1}$ i $\boldsymbol{z}_{k}$.

&nbsp;

&nbsp;

#### **Mètode de la potència iterada amb quocient de Rayleigh normalitzat**

El mètode es basa en la iteració següent:

$$
\left\{
\begin{aligned}&\boldsymbol{z}_0\in\mathbb{C}^n\setminus\{\boldsymbol{0}\},\\& \boldsymbol{z}_{k+1} = \cfrac{A\boldsymbol{z}_{k}}{\|A\boldsymbol{z}_{k}\|}, \,\,k\geq 0,\end{aligned}
\right.
$$

i en el càlcul dels termes successius $\{\sigma_k\}_{k\geq 0}$:

$$
\sigma_k = \boldsymbol{z}_k^*A\boldsymbol{z}_k,
$$

que tendiran a l'autovalor dominant $\lambda_1$.

&nbsp;

**Observació 1:** Amb aquesta variant del mètode evitem que els valors de $\boldsymbol{z}_k$ es disparin (si $|\lambda_1| > 1$) o prenguin valors molt petits (si $|\lambda_1| < 1$).

&nbsp;



##### **Remarques importants del mètode de la potència:**


*   La velocitat de convergència del mètode depèn del quocient $\frac{|\lambda_2|}{|\lambda_1|}$. Com més petit, més ràpid convergirà el mètode.



* Pel cas de matriu a coeficients reals, si $|\lambda_1| > |\lambda_2|$, necessàriament $\lambda_1$ ha de ser un valor real.

&nbsp;

#### **Mètode de la potència inversa**

Recordem que si $\lambda \neq 0$ és VAP de la matriu $A$ de VEP $\boldsymbol{v}$ aleshores, si aquesta té inversa:

$$
A\boldsymbol{v} = \lambda\boldsymbol{v} \Longleftrightarrow A^{-1}\boldsymbol{v}  = \frac{1}{\lambda}\boldsymbol{v}
$$

és a dir $\cfrac{1}{\lambda}$ és VAP d'$A^{-1}$ de vep $\boldsymbol{v}$.

Així, només cal aplicar el mètode de la potència amb la matriu $A^{-1}$ per tal d'obtenir el VAP de |A| de mòdul mínim (aplicant les mateixes condicions vistes pel Teorema i Corol·lari anterior).

Ara bé, enlloc de calcular la inversa, i considerar la iteració

$$
    \boldsymbol{z}_{k+1} = A^{-1}\boldsymbol{z}_k
$$

podem fer la descomposició de $A$ i resoldre el sistema lineal:

$$
    A\boldsymbol{z}_{k+1} = \boldsymbol{z}_{k}
$$

**Observació:** A la pràctica, també anirem normalitzant els vectors $\boldsymbol{z}_k$ d'aquest mètode:

$
\begin{aligned}
&\texttt{for } k = 0 \div n_{\text{max}}-1\\
&\phantom{for } \boldsymbol{z}_{k+1} \leftarrow A^{-1}\boldsymbol{z}_{k}\\
&\phantom{for }\sigma_k \leftarrow \boldsymbol{z}_{k}^*\boldsymbol{z}_{k+1}\\
&\phantom{for }\# \texttt{Comprovem convergència...}\\
&\phantom{for } \boldsymbol{z}_{k+1} \leftarrow \frac{\boldsymbol{z}_{k+1}}{\|\boldsymbol{z}_{k+1}\|}\\
&\texttt{end for}
\end{aligned}
$


&nbsp;

### **Què sabem fins ara?**

A la pràctica, donada una matriu $A$ quadrada amb VAPs $|\lambda_1| \geq |\lambda_1| \geq \ldots \geq|\lambda_n|$. Podem:



*   Trobar el seu VAP $\lambda_1$ estrictament dominant emprant i el seu VEP $\boldsymbol{v}_1$ associat amb el mètode de la potència.

*   Si $|\lambda_n|$ és el VAP estrictament més petit amb mòdul (amb o sense multiplicitat) i és diferent de 0, podem aplicar el mètode de la potència inversa per calcular-lo i obtenir el seu VEP $\boldsymbol{v}_n$ associat.

&nbsp;





## Variants del mètode de la potència

Fins ara ja hem vist que, utilitzant la inversa de la matriu podem calcular el VAP més petit i el seu VEP associat. Però podem fer alguna cosa més?

**Recordem també:**

##### **Teorema:**

Sigui $A$ diagonalitzable amb autovalors $\lambda_1,\lambda_2,\ldots,\lambda_n$ tals que existeix un $r\in\{1,2,\ldots,n\}$ per al qual

*   $|\lambda_1| = |\lambda_2| = \ldots = |\lambda_r| > |\lambda_{r+1}| \geq |\lambda_{r+2}| \geq \ldots \geq |\lambda_n|$,

*   $\lambda_1 = \lambda_2 = \ldots = \lambda_r$.

Aleshores, per tot $\boldsymbol{z}_0 = \alpha_1\boldsymbol{v}_1 + \alpha_2\boldsymbol{v}_2 + \ldots + \alpha_n\boldsymbol{v}_n$, amb $(\alpha_1,\ldots,\alpha_r)\in\mathbb{C}^r\setminus\{\boldsymbol{\boldsymbol{0}}\}$ podem calcular, mitjançant el mètode de la potència, el VAP $\lambda_1$ i el seu VEP associat.

&nbsp;

##### **Corol·lari:**

Considerant les mateixes hipòtesis del Teorema anterior, si $\boldsymbol{z}_0$ és tal que per algun $j\in\{r+1,r+2,\ldots,n\}$ es té:

* $\alpha_1 = \alpha_2 = \ldots = \alpha_{j-1} = 0, \quad \alpha_j \neq 0$

* $|\lambda_j| > |\lambda_{j+1}|$

Aleshores podem calcular $\lambda_j$ i el seu VEP associat amb el mètode de la potència.

&nbsp;

**Observació:** Podem generalitzar el Corol·lari anterior i fer que el $\lambda_j$ tingui multiplicitat, és a dir, existeixi un $s$ tal que $\lambda_j = \lambda_{j+1} = \ldots =\lambda_{j+s}$ i $|\lambda_{j+s}| > |\lambda_{j+s+1}|$.

&nbsp;

Ara bé, a la pràctica això no ens serveix de gaire...

&nbsp;

**Necessitem alternatives...**

Notem que si $\lambda$ és VAP de la matriu $A$ de VEP $\boldsymbol{v}$ aleshores:

$$
A\boldsymbol{v} = \lambda\boldsymbol{v} \Longleftrightarrow (A-\mu Id)\boldsymbol{v} = (\lambda -\mu)\boldsymbol{v},
$$

és a dir, $\lambda-q$ és VAP de la matriu $M = A-qId$ de vep $\boldsymbol{v}$. Seguint aquesta idea farem servir el següent mètode:

&nbsp;

### **Mètode de la potència desplaçada**

En lloc de buscar el VAP dominant $\lambda_1$ de $A$ buscarem el VAP dominant de $M$ a partir d'un valor $q$ astut, tot aplicant el mètode de la potència iterada amb quocient de Rayleigh normalitzat a aquesta matriu.

Notem que l'espectre de la matriu $M$ (el conjunt de valor propis) és el mateix que el de la matriu $A$ però desplaçat $q$:

$$\text{spec}(M) = \{\lambda_1 - q, \lambda_2 - q,\ldots,\lambda_n-q\}.$$

Observem que, depenent la distribució de l'espectre de $A$ podem arribar a calcular diversos valors propis (aplicant el mètode amb diferents valors de $q$). Ara bé, si tots el VAPs són reals, únicament podrem calcular el més gran i el més petit.

#### *Exemple*

*Sabem que la matriu*
$$
A = \begin{pmatrix}1 & -3 & 3\\ -1 & 1 & 1 \\ -1 & -3 & 5\end{pmatrix},
$$

*satisfà:*

$$\text{spec}(A) = \{1,2,4\}.$$

&nbsp;

*Així, si apliquem el mètode de la potència iterada amb quocient de Rayleigh normalitzat a la matriu $A$ obtindrem el seu VAP dominant $\lambda_1 = 4$ i el seu VEP associat.*

&nbsp;

*Si considerem ara la matriu $M = A - qId$ amb $q = 3$, tenim que el seu espectre és:*

$$\text{spec}(M) = \{-2,-1,1\}.$$

*I, per tant, si apliquem el mètode de la potència iterada amb quocient de Rayleigh normalitzat a la matriu $M$ obtindrem el seu VAP dominant -2 i el seu VEP associat. El qual correspon al VAP $\lambda = -2 +q = 1$ de la matriu $A$.*

&nbsp;

$$
\frac{|\lambda_2|}{|\lambda_1|}
$$

En particular, volem un $q$ tal que

## **Mètode de Deflació**

Suposem que treballem amb una matriu diagonalitzable.

**Objectiu:** Donat un valor i un vector propi, volem transformar la matriu de manera que ens quedi una de més senzilla. (Podríem fer l'analogia amb les arrels d'un polinomi, un cop trobada una arrel treballem amb un polinomi més senzill).

Així doncs, intentarem reduir la nostra matriu $n\times n$ a una matriu de mides $(n-1)\times(n-1)$. Ens centrarem en el cas real, però aquest resultat es poden estendre al cas complex.

Per fer-ho necessitarem recordar dos conceptes que ja coneixem:

* El concepte de matrius semblants.

* Les matrius de Householder (teniu més detall al Jupyter notebook 2g).

Recordem-los ràpidament:




##### **Definició:** Matrius semblants

Donades dues matrius $A$ i $B$ de dimensió $n\times n$ direm que són semblants si existeix alguna matriu invertible $P$ tal que:

$$B=P^{-1}AP$$

**Exercici:**

Demostreu que dues matrius semblants tenen mateix VAPs:

Tenim:

$$Av = \lambda v$$
$$B = P^{-1}AP$$

Intentem que les coses s'assemblin:

$$ P^{-1}Av = \lambda P^{-1} v$$
$$ P^{-1}A · Id ·v = \lambda P^{-1} v$$
$$ P^{-1}A\underbrace{PP^{-1}}_{Id}v = \lambda P^{-1} v$$
$$ BP^{-1}v = \lambda P^{-1} v$$

Ens queda que $P^{-1}v$ es vep del mateix vap que A :)


##### **Propietats:** de les matrius semblants

Si $\boldsymbol{v}$ és un vector propi de $A$ amb valor propi $\lambda$, aleshores $P^{-1} \boldsymbol{v}$ és un vector propi de $B$ amb valor propi $\lambda$.

##### **Definició:** Matriu de Householder:

Per cada vector no nul $\boldsymbol{u}\in\mathbb{R}^n$ definim la seva matriu de Householder associada com:

$$
    P(\boldsymbol{u}) := Id - \frac{2}{\boldsymbol{u}^\top \boldsymbol{u}} \boldsymbol{u} \boldsymbol{u}^\top.
$$

Fixem-nos que $\boldsymbol{u}$ és un vector columna i, per tant,


*   $\boldsymbol{u}^\top \boldsymbol{u} = \boldsymbol{u}\cdot \boldsymbol{u} = \|\boldsymbol{u}\|^2$ és un escalar.

*   $\boldsymbol{u} \boldsymbol{u}^\top$ és una matriu simètrica:

$$
\boldsymbol{u} \boldsymbol{u}^\top =
\begin{pmatrix}
u_1^2 & u_1 u_2 & \cdots & u_1 u_n\\
u_1 u_2 & u_2^2 & \cdots & u_2 u_n\\
\vdots & \vdots & \ddots & \vdots\\
u_1 u_n & u_2 u_n & \cdots & u_n^2\\
\end{pmatrix}.
$$

##### **Propietats:** de les Matrius de Householder

Sigui $\boldsymbol{u}\in\mathbb{R}^n\setminus\{\boldsymbol{0}\}$. Aleshores:

1.   $P(k \boldsymbol{u}) = P(\boldsymbol{u}), \forall k \neq 0$.

2.  $P(\boldsymbol{u})\boldsymbol{u} = - \boldsymbol{u}$

3.   $P(\boldsymbol{u})$ és una matriu simètrica ($P(\boldsymbol{u}) = P(\boldsymbol{u})^\top$).

4.   $P(\boldsymbol{u})P(\boldsymbol{u}) = Id$

5.   $P(\boldsymbol{u}) = P(\boldsymbol{u})^{-1} = P(\boldsymbol{u})^\top$, és a dir, $P(\boldsymbol{u})$ és una matriu ortonormal.

6.   $\det(P(\boldsymbol{u})) = -1$.

&nbsp;

##### **Transformació entre dos vectors amb el mateix mòdul via matrius de Householder**

Suposem que tenim els vector $\boldsymbol{a},\,\boldsymbol{b}\in\mathbb{R}^n\setminus\{\boldsymbol{0}\}$ tal que $\|\boldsymbol{a}\| = \|\boldsymbol{b}\|$. Podem passar d'un a l'altre via una transformació de Householder:

$$\begin{aligned}&\boldsymbol{a} = P(\boldsymbol{u})\boldsymbol{b}\\&\boldsymbol{b} = P(\boldsymbol{u})\boldsymbol{a}\end{aligned}$$

definint $\boldsymbol{u} = \pm(\boldsymbol{a} - \boldsymbol{b})$.

&nbsp;

D'aquesta forma, si $\lambda_1$ és VAP de $A$ i suposem que $\boldsymbol{v}$ és el seu VEP associat, el qual suposarem normalitzat, de manera que $\|\boldsymbol{v}\| = 1$, volem construir una matriu de Householder $H$ de manera que

$$
    H\boldsymbol{v} = \pm \boldsymbol{e}_1,
$$

on $\boldsymbol{e}_1$ és el vector canònic.

&nbsp;

> **Amb detall:**
>
> Tenim $A\boldsymbol{v} = \lambda_1\boldsymbol{v}$, és a dir $\boldsymbol{v}$ així, si multipliquem per una matriu de Householder, tenim:
>
> $$\begin{aligned}A\boldsymbol{v} = \lambda_1\boldsymbol{v} & \Longleftrightarrow HA\boldsymbol{v} = HA\underbrace{H^{-1}H}_{Id}\boldsymbol{v} = H\lambda_1\boldsymbol{v} = \lambda_1H\boldsymbol{v}\end{aligned}$$
>
> És a dir, $H\boldsymbol{v}$ és VEP de VAP $\lambda_1$ de la matriu $HAH^{-1}$.

&nbsp;




Recordem que tenim llibertat per triar el signe del vector, i el triarem de manera que poguem evitar cancel·lacions, és a dir, signe oposat a la primera coordenada de $\boldsymbol{v}$. És a dir, definim:

$$
\boldsymbol{u} = \begin{pmatrix}v_1 +\text{sign}(v_1)\\v_2\\\vdots\\v_n\end{pmatrix},
$$

i considerarem la matriu de Householder:

$$
H = P(\boldsymbol{u}).
$$

&nbsp;

D'aquesta forma, transformarem la matriu $A$ en una matriu semblant:

$$
A_1 = H A H^{-1} = \begin{pmatrix}\lambda_1 & \boldsymbol{b}^\top\\ \boldsymbol{0} & \tilde{A}_1\end{pmatrix}
$$

la qual té els mateixos valors propis, però ara hem imposat que el seu VEP de VAP $\lambda_1$ sigui el vector canònic $\pm\boldsymbol{e}_1$.

&nbsp;

La idea ara és que treballarem amb $\tilde{A}_1$ i podem tornar a calcular un VAP i el seu VEP associat i anar repetint el procediment.

&nbsp;



**Pregunta:** I si calculem un VAP $\lambda_2$ i el seu VEP associat de $\tilde{A}_1$ com recuperm el VEP original de $A$?

&nbsp;

Subposem $\tilde{\boldsymbol{w}}_2$ VEP de VAP $\lambda_2$ de $\tilde{A}_1$. Sabem que satisfà:

$$
\tilde{A}_1\tilde{\boldsymbol{w}}_2 = \lambda_2 \tilde{\boldsymbol{w}}_2
$$

Intentem trobar el VEP $\boldsymbol{w}_2$ de VAP $\lambda_2$ de $A_1$. Sabem que ha de satisfer:

$$
\begin{pmatrix}\lambda_1 & \boldsymbol{b}^\top\\ \boldsymbol{0} & \tilde{A}_1\end{pmatrix} \underbrace{\begin{pmatrix} \alpha \\ \tilde{\boldsymbol{w}}_2 \end{pmatrix}}_{\boldsymbol{w}_2} = \lambda_2 \begin{pmatrix} \alpha \\ \tilde{\boldsymbol{w}}_2 \end{pmatrix}
$$

per algun $\alpha$.

En particular, ha de satisfer:

$$
    \alpha \lambda_1 + \boldsymbol{b}^\top\tilde{\boldsymbol{w}}_2 = \alpha\lambda_2
$$

Notem que si $\lambda_1 = \lambda_2$ podem prendre directament $\alpha = 0$, en cas $\lambda_1 \neq \lambda_2$, prendrem:

$$
    \alpha = \frac{\boldsymbol{b}^\top\tilde{\boldsymbol{w}}_2}{\lambda_2-\lambda_1}.
$$

&nbsp;

Ara, un cop determinat $\boldsymbol{w}_2$ podem trobar $\boldsymbol{v}_2$, el vector propi de VAP $\lambda_2$ de $A$ calculant:

$$
    \boldsymbol{v}_2 = H^{-1}\boldsymbol{w}_2.
$$

#### Resolució:

És fàcil veure que la matriu $A$ té VAP $\lambda_1$ i VEP associat $\boldsymbol{v}_1 = \boldsymbol{e}_1$.

Així doncs,

$$A_1 = A,$$

ja que la matriu de Householder seria la identitat ($H_1=Id$).

Centrem-nos ara doncs en la submatriu

$$\tilde{A}_1 = \begin{pmatrix} 4 & 1\\ -6 & -3\end{pmatrix}.$$

Suposem que apliquem el mètode de la potència i arribem a

$$\lambda_2 = 3$$

i

$$\tilde{\boldsymbol{w}}_2 = \begin{pmatrix}\frac{\sqrt{2}}{2}\\-\frac{\sqrt{2}}{2}\end{pmatrix}.$$

Calculem ara $\boldsymbol{w}_2$, és a dir, determinem el valor de $\alpha$.

Com $\lambda_1 = 1 \neq 3 = \lambda_2$, tenim:

$$\alpha = \frac{\boldsymbol{b}^\top\tilde{\boldsymbol{w}}_2}{\lambda_2-\lambda_1} = \frac{1}{3-1}\begin{pmatrix} 4 & 3\end{pmatrix}\begin{pmatrix}\frac{\sqrt{2}}{2}\\-\frac{\sqrt{2}}{2}\end{pmatrix} = \frac{\sqrt{2}}{4}$$

i, per tant, tenim:

$$\boldsymbol{w}_2 = \begin{pmatrix}\frac{\sqrt{2}}{4}\\ \frac{\sqrt{2}}{2}\\-\frac{\sqrt{2}}{2}\end{pmatrix}$$

o, multiplicant per $2\sqrt{2}$:

$$\boldsymbol{w}_2 = \begin{pmatrix}1\\ 2\\-2\end{pmatrix}$$

Notem que, en aquest cas, com $H=Id$ tenim que $\boldsymbol{v}_2 = \boldsymbol{w}_2$.

&nbsp;

&nbsp;

Ara, podríem continuar el procediment a partir de la matriu $B = \tilde{A}_1$ i aplicar deflació, ja que coneixem un VAP i el seu VEP associat:

$$\lambda_2 = 3$$

i

$$\boldsymbol{\omega}_2 = \begin{pmatrix}\frac{\sqrt{2}}{2}\\-\frac{\sqrt{2}}{2}\end{pmatrix}.$$

Calculem el vector $\boldsymbol{u}$ (notem que $\|\boldsymbol{\omega}_2\| = 1$):

$$\boldsymbol{u} = \omega_2 + \text{sign}(u_1)\boldsymbol{e}_1 = \begin{pmatrix}\frac{\sqrt{2}}{2} + 1\\-\frac{\sqrt{2}}{2}\end{pmatrix}$$

D'aquesta forma, la matriu de Householder associada a $\boldsymbol{u}$ esdevé:

$$H = \begin{pmatrix}-\frac{\sqrt{2}}{2} & \frac{\sqrt{2}}{2}\\ \frac{\sqrt{2}}{2} & \frac{\sqrt{2}}{2}\end{pmatrix}$$

i, tenim:

$$B_1 = H B H^{-1} = \begin{pmatrix} 3 & -7\\ 0 & -2 \end{pmatrix},$$

on ja veiem clarament que l'altre valor propi serà $\lambda_3 = -2$.

Per calcular el VEP $\boldsymbol{\omega}_3$ de VAP $\lambda_3 = -2$, hauríem de resoldre:

$$ \alpha = \frac{\boldsymbol{b}^\top\tilde{\boldsymbol{w}}_3}{\lambda_3-\lambda_2} = \frac{1}{-2-3}\left(-7 \cdot 1\right) = \frac{7}{5}$$

Així doncs,

$$\boldsymbol{\omega}_3 = \begin{pmatrix} \frac{7}{5}\\ 1\end{pmatrix}$$

és VEP de $B_1$ de VAP $-2$.

Ara, per trobar el VEP de $B$ multiplicarem per la matriu $H^{-1}$ anterior:

$$\tilde{\boldsymbol{w}}_3 = H^{-1}\boldsymbol{\omega}_3 = \begin{pmatrix}-\frac{\sqrt{2}}{2} & \frac{\sqrt{2}}{2}\\ \frac{\sqrt{2}}{2} & \frac{\sqrt{2}}{2}\end{pmatrix}\begin{pmatrix} \frac{7}{5}\\ 1\end{pmatrix} = \left(\begin{array}{c} -\frac{\sqrt{2}}{5}\\ \frac{6\,\sqrt{2}}{5} \end{array}\right).$$

I, un cop tenim el VEP de $B = \tilde{A}_1$ repetim l'operació per determinar $\alpha$ i obtenir així el VEP de $A_1$:

$$
    \alpha = \frac{\boldsymbol{b}^\top\tilde{\boldsymbol{w}}_3}{\lambda_3-\lambda_1} = \frac{1}{-2-1}\begin{pmatrix} 4 & 3\end{pmatrix}\left(\begin{array}{c} -\frac{\sqrt{2}}{5}\\ \frac{6\,\sqrt{2}}{5} \end{array}\right) = \frac{\sqrt{2}}{4} = -\frac{14\sqrt{2}}{15}.
$$

Així:

$$\boldsymbol{w}_3 = \begin{pmatrix}-\frac{14\sqrt{2}}{15}\\ -\frac{\sqrt{2}}{5}\\ \frac{6\,\sqrt{2}}{5}\end{pmatrix}$$

o, multiplicant per $\frac{15}{\sqrt{2}}$:

$$\boldsymbol{w}_3 = \begin{pmatrix}-14\\ -3\\ 18\end{pmatrix}$$

I, com $A=A_1$, tenim que $\boldsymbol{v}_3 = \boldsymbol{w}_3$.
